```python
#
# @lc app=leetcode.cn id=144 lang=python
#
# [144] 二叉树的前序遍历
#


# @lc code=start
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        r = []  # 整体思想 递归

        def dfs(root):
            if root == None:
                return
            r.append(root.val)
            dfs(root.left)
            dfs(root.right)

        dfs(root)
        return r


# @lc code=end
```

```python
#
# @lc app=leetcode.cn id=144 lang=python
#
# [144] 二叉树的前序遍历
#


# @lc code=start
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        stack = []
        res = []
        while stack or root:
            while root:
                res.append(root.val)
                stack.append(root)
                root = root.left
            root = stack.pop()
            root = root.right
        return res


# @lc code=end
```

```python
#
# @lc app=leetcode.cn id=144 lang=python
#
# [144] 二叉树的前序遍历
#


# @lc code=start
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object): # 前中后栈方法模板
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        stack = []
        res = []
        if root: stack.append(root)
        while stack: # 栈空结束
            root = stack.pop() # 弹出判断是否是None
            if root != None:
                if root.right: stack.append(root.right) # 前序是中左右 所以入栈是右左中
                if root.left: stack.append(root.left) # 添加左节点（空节点不入栈）
                stack.append(root) # 添加中节点
                stack.append(None) # 中节点访问过，但是还没有处理，加入空节点做为标记。
            else: # 是None，则再弹出一个，并将值放入res
                root = stack.pop()
                res.append(root.val)
        return res


# @lc code=end
```
