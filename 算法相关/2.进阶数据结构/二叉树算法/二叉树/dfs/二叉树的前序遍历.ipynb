```python
#
# @lc app=leetcode.cn id=144 lang=python
#
# [144] 二叉树的前序遍历 中左右
#


# @lc code=start
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
# 递归法
class Solution(object):
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        res = []

        def dfs(root):
            if root == None: return
            res.append(root.val)
            dfs(root.left)
            dfs(root.right)

        dfs(root)
        return res


# @lc code=end
```

```python
# 中左右
# 迭代往左 往左过程入栈并添加结果
class Solution(object):
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        res = []
        stack = []
        while stack or root: # 当stack空 且 root空 就退出
            while root:
                res.append(root.val)
                stack.append(root)
                root = root.left
            root = stack.pop()  # 当root空跳出时，stack还保存着上一个非空的节点
            root = root.right
        return res
```

```python
# 中左右
# 反序入栈判断
class Solution(object):
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        res = []
        stack = []
        if root: stack.append(root)
        while stack:
            root = stack.pop() # 弹出判断
            if root: # stack能pop说明root必定存在 可以取消if root
                res.append(root.val) # 写哪里都一样
                if root.right: stack.append(root.right) # 右左顺序入栈 出栈就是左右
                if root.left: stack.append(root.left)
        return res
```

```python
# 中左右
# 前中后通用迭代模板 标记法 入栈 右左中
class Solution(object):
    def preorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        res = []
        stack = []
        if root: stack.append(root)
        while stack:
            root = stack.pop()  # 弹出判断
            if root:
                if root.right: stack.append(root.right)  # 右左顺序入栈 出栈就是左右
                if root.left: stack.append(root.left)
                stack.append(root)
                stack.append(None)
            else:
                root = stack.pop()  # None判断前已经弹出  继续弹出被标记的加入res
                res.append(root.val)
        return res
```
