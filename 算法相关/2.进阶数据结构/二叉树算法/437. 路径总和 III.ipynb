给定一个二叉树的根节点 root ，和一个整数 targetSum ，求该二叉树里节点值之和等于 targetSum 的 路径 的数目。

路径 不需要从根节点开始，也不需要在叶子节点结束，但是路径方向必须是向下的（只能从父节点到子节点）。

示例 1：
输入：root = [10,5,-3,3,2,null,11,3,-2,null,1], targetSum = 8
输出：3
解释：和等于 8 的路径有 3 条，如图所示。

示例 2：
输入：root = [5,4,8,11,null,13,4,7,2,null,null,5,1], targetSum = 22
输出：3

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# 双递归
class Solution:
    def pathSum(self, root: TreeNode, targetSum: int) -> int:
        if root is None: return 0
        cnt = 0

        def dfs(node: TreeNode, Sum: int): # 经过node节点的符合条件的个数
            nonlocal cnt
            if Sum == targetSum:
                cnt += 1
            if node.left:
                dfs(node.left, Sum + node.left.val)
            if node.right:
                dfs(node.right, Sum + node.right.val)

        def dfs2(node: TreeNode): # 递归所有 经过该节点
            dfs(node, node.val)
            if node.left:
                dfs2(node.left)
            if node.right:
                dfs2(node.right)

        # dfs(root, root.val)
        dfs2(root)
        return cnt
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


from collections import defaultdict


# 前缀和
class Solution:
    def pathSum(self, root: TreeNode, targetSum: int) -> int:
        if root is None: return 0
        prefix = defaultdict(int)
        prefix[0] = 1
        cnt = 0

        def dfs(node: TreeNode, cur_sum: int):
            nonlocal cnt
            cur_sum += node.val
            cnt += prefix[cur_sum - targetSum]
            for n in [node.left, node.right]:
                if n:
                    prefix[cur_sum] += 1
                    dfs(n, cur_sum)
                    prefix[cur_sum] -= 1

        dfs(root, 0)
        return cnt
```

```python
from collections import defaultdict


class Solution:
    def pathSum(self, root: TreeNode, targetSum: int) -> int:
        if not root:
            return 0
        ans = 0
        # 生成记录前缀和的字典
        prefix = defaultdict(int)
        prefix[0] = 1  # 表示根节点的前缀和为0

        def dfs(root: TreeNode, cursum):
            nonlocal ans
            if not root:
                return
            cursum += root.val  # 前缀和加上当前数值
            # 算算当前路径下能构成target的前缀和个数
            ans += prefix[cursum - targetSum]
            # 回溯递归求解
            prefix[cursum] += 1
            dfs(root.left, cursum)
            dfs(root.right, cursum)
            prefix[cursum] -= 1

        dfs(root, 0)
        return ans
```
