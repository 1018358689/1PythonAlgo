```python
# 线段树
from typing import List


class NumArray:

    def __init__(self, nums: List[int]):
        self.sz = len(nums)
        self.SegmentTree = [0] * self.sz + nums
        for i in range(self.sz - 1, 0, -1):
            self.SegmentTree[i] = self.SegmentTree[2 * i] + self.SegmentTree[2 * i + 1]

    def update(self, index: int, val: int) -> None:
        i = index + self.sz
        d = val - self.SegmentTree[i]
        while i != 0:
            self.SegmentTree[i] += d
            i //= 2

    def sumRange(self, left: int, right: int) -> int:
        l, r = left + self.sz, right + self.sz
        total = 0
        while l <= r:
            if l % 2 == 1: # 左区间奇数说明是右孩子 结果包含左区间但不包含其父节点
                total += self.SegmentTree[l] # 那么结果里先将该孩子加入 然后i+1转向下一个区间
                l += 1 # 左区间变成i+1 必然是偶数 说明结果包含其父节点
            if r % 2 == 0: # 右区间偶数说明是左孩子 结果包含右区间但不包含其父节点
                total += self.SegmentTree[r]
                r -= 1
            l //= 2
            r //= 2
        return total


# Your NumArray object will be instantiated and called as such:
# obj = NumArray(nums)
# obj.update(index,val)
# param_2 = obj.sumRange(left,right)
```

```python
from typing import List
# 超时

class NumArray:

    def __init__(self, nums: List[int]):
        self.nums = nums
        self.preSum = [0] * (len(self.nums) + 1)
        for i in range(len(self.nums)):
            self.preSum[i + 1] = self.preSum[i] + self.nums[i]

    def update(self, index: int, val: int) -> None:
        for i in range(index + 1, len(self.preSum)):
            self.preSum[i] += (val - self.nums[index])
        self.nums[index] = val

    def sumRange(self, left: int, right: int) -> int:
        return self.preSum[right + 1] - self.preSum[left]


# Your NumArray object will be instantiated and called as such:
# obj = NumArray(nums)
# obj.update(index,val)
# param_2 = obj.sumRange(left,right)
```

```python
from typing import List
# 超时

class NumArray:

    def __init__(self, nums: List[int]):
        self.nums = nums

    def update(self, index: int, val: int) -> None:
        self.nums[index] = val

    def sumRange(self, left: int, right: int) -> int:
        return sum(self.nums[left:right + 1])


# Your NumArray object will be instantiated and called as such:
# obj = NumArray(nums)
# obj.update(index,val)
# param_2 = obj.sumRange(left,right)
```

```python
from typing import List


class NumArray:

    def __init__(self, nums: List[int]):
        self.nums = nums
        self.total = sum(self.nums)

    def update(self, index: int, val: int) -> None:
        self.total += val - self.nums[index]
        self.nums[index] = val
        

    def sumRange(self, left: int, right: int) -> int:
        return self.total - sum(self.nums[:left]) - sum(self.nums[right + 1:])


# Your NumArray object will be instantiated and called as such:
# obj = NumArray(nums)
# obj.update(index,val)
# param_2 = obj.sumRange(left,right)
```
