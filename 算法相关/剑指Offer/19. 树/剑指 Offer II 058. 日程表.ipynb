```python
class Node:
    def __init__(self, si, ei):
        self.si = si
        self.ei = ei
        self.mi = si + ei >> 1
        self.val = 0  # 区间和
        self.lazy = 0  # 未下推
        self.left = None
        self.right = None


class SegmentTree:
    def __init__(self, si, ei):
        self.root = Node(si, ei)

    def pushDown(self, node):
        if not node.left: node.left = Node(node.si, node.mi)
        if not node.right: node.right = Node(node.mi + 1, node.ei)
        if node.lazy == 0:
            return
        node.left.val += node.lazy * (node.left.ei - node.left.si + 1)
        node.right.val += node.lazy * (node.right.ei - node.right.si + 1)
        node.left.lazy += node.lazy
        node.right.lazy += node.lazy
        node.lazy = 0

    def update(self, node, l, r, val):  # [l,r]每个值加val
        if node.ei < l or node.si > r:
            return
        if l <= node.si <= node.ei <= r:
            node.val += val * (node.ei - node.si + 1)
            node.lazy += val
            return
        self.pushDown(node)
        if node.ei >= l:
            self.update(node.left, l, r, val)
        if node.mi + 1 <= r:
            self.update(node.right, l, r, val)
        node.val = node.left.val + node.right.val

    def query(self, node, l, r):
        if node.ei < l or node.si > r:
            return 0
        if l <= node.si <= node.ei <= r:
            return node.val
        self.pushDown(node)
        ans = 0
        if node.ei >= l:
            ans += self.query(node.left, l, r)
        if node.mi + 1 <= r:
            ans += self.query(node.right, l, r)
        return ans


class MyCalendar:
    def __init__(self):
        self.st = SegmentTree(0, 10**9)

    def book(self, start: int, end: int) -> bool:
        res = self.st.query(self.st.root, start, end - 1)
        if res:
            return False
        else:
            self.st.update(self.st.root, start, end - 1, 1)
            return True


# Your MyCalendar object will be instantiated and called as such:
# obj = MyCalendar()
# param_1 = obj.book(start,end)
```
