```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None


class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        ans = ''

        def dfs(root):
            nonlocal ans
            if not root:
                ans += '#,'
                return
            ans += f'{root.val},'
            dfs(root.left)
            dfs(root.right)

        dfs(root)
        return ans

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        data = data.split(',')[:-1]

        def dfs(data):
            nonlocal i
            if data[i] == '#':
                return None
            root = TreeNode(data[i])
            i += 1
            root.left = dfs(data)
            i += 1
            root.right = dfs(data)
            return root

        i = 0
        return dfs(data)


# Your Codec object will be instantiated and called as such:
# ser = Codec()
# deser = Codec()
# ans = deser.deserialize(ser.serialize(root))
```
