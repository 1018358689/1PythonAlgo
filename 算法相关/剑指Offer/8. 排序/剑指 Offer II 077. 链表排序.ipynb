```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def sortList(self, head: ListNode) -> ListNode:
        nums = []
        while head:
            nums.append(head.val)
            head = head.next
        nums.sort()
        dummy = ListNode()
        p = dummy
        for n in nums:
            p.next = ListNode()
            p.next.val = n
            p = p.next
        return dummy.next
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def sortList(self, head: ListNode) -> ListNode:
        def merge(l1: ListNode, l2: ListNode):
            dummy = ListNode()
            p = dummy
            while l1 and l2:
                p.next = ListNode()
                if l1.val <= l2.val:
                    p.next.val = l1.val
                    l1 = l1.next
                else:
                    p.next.val = l2.val
                    l2 = l2.next
                p = p.next
            p.next = l1 if l1 else l2
            return dummy.next

        def dc(head: ListNode):
            if not head or not head.next:
                return head
            pre, slow, fast = None, head, head
            while fast and fast.next:
                pre = slow
                slow = slow.next
                fast = fast.next.next
            pre.next = None
            left = dc(head)
            right = dc(slow)
            return merge(left, right)

        return dc(head)
```
