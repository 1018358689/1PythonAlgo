```python
class RecentCounter:

    def search(self, nums, target):  # >=target
        l, r = 0, len(nums) - 1
        while l <= r:
            m = l + r >> 1
            if nums[m] >= target:
                r = m - 1
            else:
                l = m + 1
        return l

    def __init__(self):
        self.q = []
        self.lastIndex = -1

    def ping(self, t: int) -> int:
        self.q.append(t)
        self.lastIndex += 1
        return self.lastIndex - self.search(self.q, t - 3000) + 1


# Your RecentCounter object will be instantiated and called as such:
# obj = RecentCounter()
# param_1 = obj.ping(t)
```

```python
from collections import deque


class RecentCounter:

    def __init__(self):
        self.q = deque([])

    def ping(self, t: int) -> int:
        self.q.append(t)
        while self.q[0] < t - 3000:
            self.q.popleft()
        return len(self.q)


# Your RecentCounter object will be instantiated and called as such:
# obj = RecentCounter()
# param_1 = obj.ping(t)
```

```python
# 超时 练习线段树
class Node:

    def __init__(self, si, ei):  # 节点范围[si,ei]
        self.si = si
        self.ei = ei
        self.mi = si + ei >> 1
        self.val = 0  # 区间[si,ei]的和
        self.lazy = 0  # 0/1 0:无变化 1:该节点设为1 子节点未设为1
        self.left = None
        self.right = None


class SengmentTree:

    def __init__(self, si, ei):  # 整棵树范围[si,ei]
        self.root = Node(si, ei)

    def pushDown(self, node: Node):  # 下推node的lazy标记并动态开点
        if not node.left: node.left = Node(node.si, node.mi)
        if not node.right: node.right = Node(node.mi + 1, node.ei)
        if node.lazy == 0:
            return
        node.left.val = (node.left.ei - node.left.si + 1) * node.val
        node.left.lazy = node.lazy
        node.right.val = (node.right.ei - node.right.si + 1) * node.val
        node.right.lazy = node.lazy
        node.lazy = 0

    def update(self, node: Node, l, r, val):  # 更新node的[l,r]范围值为val
        if node.ei < l or node.si > r:
            return
        if l <= node.si <= node.ei <= r:
            node.val = (node.ei - node.si + 1) * val
            node.lazy = val
            return
        self.pushDown(node)
        if node.mi >= l:
            self.update(node.left, l, r, val)
        if node.mi + 1 <= r:
            self.update(node.right, l, r, val)
        node.val = node.left.val + node.right.val

    def query(self, node: Node, l, r):  # 查询node的[l,r]范围和
        if node.ei < l or node.si > r:
            return 0
        if l <= node.si <= node.ei <= r:
            return node.val
        self.pushDown(node)
        ans = 0
        if node.mi >= l:
            ans += self.query(node.left, l, r)
        if node.mi + 1 <= r:
            ans += self.query(node.right, l, r)
        return ans


class RecentCounter:

    def __init__(self):
        self.st = SengmentTree(1, 10**9)

    def ping(self, t: int) -> int:
        self.st.update(self.st.root, t, t, 1)
        return self.st.query(self.st.root, max(1, t - 3000), t)


# Your RecentCounter object will be instantiated and called as such:
obj = RecentCounter()
obj.ping(1)
```
