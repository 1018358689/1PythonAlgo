```python
class Node:

    def __init__(self, key=-1, val=-1):
        self.key = key
        self.val = val
        self.pre = None
        self.next = None


class DoubleList:

    def __init__(self):
        self.head = Node()
        self.end = Node()
        self.head.next = self.end
        self.end.pre = self.head

    def addNodeAfter(self, node: Node, preNode: Node):
        node.next = preNode.next
        node.pre = preNode
        preNode.next.pre = node
        preNode.next = node

    def removeNode(self, node: Node):
        node.pre.next = node.next
        node.next.pre = node.pre

    def move2last(self, node: Node):
        if node == self.end.pre:
            return
        self.removeNode(node)
        self.addNodeAfter(node, self.end.pre)


class LRUCache:

    def __init__(self, capacity: int):
        self.capacity = capacity
        self.dl = DoubleList()
        self.key2node = {}

    def get(self, key: int) -> int:
        node = self.key2node.get(key)
        if not node:
            return -1
        tmp = node.val
        self.dl.move2last(node)
        return tmp

    def put(self, key: int, value: int) -> None:
        node = self.key2node.get(key)
        if not node:  # 节点不存在 需要建立节点
            if self.capacity == 0:  # 满容量 需要删除头
                headKey = self.dl.head.next.key
                self.key2node.pop(headKey)
                self.dl.removeNode(self.dl.head.next)
            else:  # 未满容量加入新节点
                self.capacity -= 1
            newNode = Node(key, value)
            self.key2node[key] = newNode
            self.dl.addNodeAfter(newNode, self.dl.end.pre)
        else:
            node.val = value
            self.dl.move2last(node)


# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)
```
