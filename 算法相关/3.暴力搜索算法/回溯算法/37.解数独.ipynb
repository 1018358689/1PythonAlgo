编写一个程序，通过填充空格来解决数独问题。

数独的解法需 遵循如下规则：

数字 1-9 在每一行只能出现一次。
数字 1-9 在每一列只能出现一次。
数字 1-9 在每一个以粗实线分隔的 3x3 宫内只能出现一次。（请参考示例图）
数独部分空格内已填入了数字，空白格用 '.' 表示。


示例：


输入：board = [
["5","3",".",".","7",".",".",".","."],
["6",".",".","1","9","5",".",".","."],
[".","9","8",".",".",".",".","6","."],
["8",".",".",".","6",".",".",".","3"],
["4",".",".","8",".","3",".",".","1"],
["7",".",".",".","2",".",".",".","6"],
[".","6",".",".",".",".","2","8","."],
[".",".",".","4","1","9",".",".","5"],
[".",".",".",".","8",".",".","7","9"]]
输出：[
["5","3","4","6","7","8","9","1","2"],
["6","7","2","1","9","5","3","4","8"],
["1","9","8","3","4","2","5","6","7"],
["8","5","9","7","6","1","4","2","3"],
["4","2","6","8","5","3","7","9","1"],
["7","1","3","9","2","4","8","5","6"],
["9","6","1","5","3","7","2","8","4"],
["2","8","7","4","1","9","6","3","5"],
["3","4","5","2","8","6","1","7","9"]]
解释：输入的数独如上图所示，唯一有效的解决方案如下所示：

提示：

board.length == 9
board[i].length == 9
board[i][j] 是一位数字或者 '.'
题目数据 保证 输入数独仅有一个解

```python
'''
算法的核心思路非常非常的简单，就是对每一个空着的格子穷举 1 到 9，
如果遇到不合法的数字（在同一行或同一列或同一个 3×3 的区域中存在相同的数字）则跳过，
如果找到一个合法的数字，则继续穷举下一个空格子。'''
def f(board: list):
    def check(board: list, row: int, col: int, content: str):
        # 检查行
        for c in range(9):
            if board[row][c] == content:
                return False
        # 检查列
        for r in range(9):
            if board[r][col] == content:
                return False
        # 检查九宫格
        start_r = row // 3 * 3
        start_c = col // 3 * 3
        end_r = start_r + 2
        end_c = start_c + 2
        for r in range(start_r, end_r + 1):
            for c in range(start_c, end_c + 1):
                if board[r][c] == content:
                    return False
        return True

    def dfs(board: list):
        for r in range(9):
            for c in range(9):
                if board[r][c] != '.':  # 已经有数字则跳过
                    continue
                for content in range(1, 10):
                    if check(board, r, c, str(content)) == False:  # 检查不通过则跳过
                        continue
                    board[r][c] = str(content)
                    if dfs(board):  # 找到一个解就返回
                        return True
                    else:
                        board[r][c] = '.'  # 当前格子全部数字试了都不行 回溯
                return False  # 当前格子全部数字试了都不行
        return True

    dfs(board)
    return board
```

```python
def f(board: list):
    def check(board: list, row: int, col: int, content: str):
        # 检查行
        for c in range(9):
            if board[row][c] == content:
                return False
        # 检查列
        for r in range(9):
            if board[r][col] == content:
                return False
        # 检查九宫格
        start_r = row // 3 * 3
        start_c = col // 3 * 3
        end_r = start_r + 2
        end_c = start_c + 2
        for r in range(start_r, end_r + 1):
            for c in range(start_c, end_c + 1):
                if board[r][c] == content:
                    return False
        return True

    def dfs(board: list, row: int, col: int):
        if row == 9:  # 找到一个可行解，触发 base case
            return True
        if col == 9:  # 穷举到最后一列的话就换到下一行重新开始。
            return dfs(board, row + 1, 0)
        if board[row][col] != '.':  # 如果有预设数字，不用我们穷举
            return dfs(board, row, col + 1)
        for c in range(1, 10):
            if check(board, row, col, str(c)) == False:  # 如果遇到不合法的数字，就跳过
                continue
            board[row][col] = str(c)
            if dfs(board, row, col + 1):  # 如果找到一个可行解，立即结束
                return True
            board[row][col] = '.'
        return False  # 穷举完 1~9，依然没有找到可行解，此路不通

    dfs(board, 0, 0)
    return board


board = [
[".",".",".",".",".",".","2","3","4"],
[".","1",".",".",".",".",".",".","."],
[".",".",".",".",".",".","5","6","."],
[".",".",".",".",".",".",".",".","."],
[".",".",".",".",".",".",".",".","."],
[".",".",".",".",".",".",".",".","."],
[".","8",".",".",".",".",".",".","."],
[".",".",".","7","9",".",".",".","."],
[".",".",".",".",".",".",".","8","."]]
f(board)
```
