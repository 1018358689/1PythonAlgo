```python
# 超时
class Node:

    def __init__(self, si, ei) -> None:
        self.si = si
        self.ei = ei
        self.mi = si + ei >> 1
        self.lgh = ei - si + 1
        self.valMax = float('-inf')  # 当前区间的最大值
        self.valMin = float('inf')  # 当前区间的最小值
        self.left = None
        self.right = None
        self.lazy = 0


class SegmentTree:

    def __init__(self, si, ei) -> None:
        self.root = Node(si, ei)

    def pushDown(self, node: Node):
        if not node.left: node.left = Node(node.si, node.mi)
        if not node.right: node.right = Node(node.mi + 1, node.ei)
        if not node.lazy:
            return
        node.left.valMax = node.lazy  # 往下传 大区间设为lazy 子区间一样要设
        node.left.valMin = node.lazy
        node.right.valMax = node.lazy
        node.right.valMin = node.lazy
        node.left.lazy = node.lazy
        node.right.lazy = node.lazy
        node.lazy = 0

    def update(self, node: Node, l, r, val):  # 给[l,r]区间每个值设为val
        if r < node.si or node.ei < l:
            return
        if l <= node.si <= node.ei <= r:
            node.valMax = val
            node.valMin = val
            node.lazy = val
            return
        self.pushDown(node)
        if l <= node.left.ei:
            self.update(node.left, l, r, val)
        if node.right.si <= r:
            self.update(node.right, l, r, val)
        node.valMax = max(node.left.valMax, node.right.valMax)
        node.valMin = min(node.left.valMin, node.right.valMin)

    def queryMax(self, node: Node, l, r):  # 查询[l,r]区间的最大值
        if r < node.si or node.ei < l:
            return float('-inf')
        if l <= node.si <= node.ei <= r:
            return node.valMax
        self.pushDown(node)
        ans = float('-inf')
        if l <= node.left.ei:
            ans = max(ans, self.queryMax(node.left, l, r))
        if node.right.si <= r:
            ans = max(ans, self.queryMax(node.right, l, r))
        return ans

    def queryMin(self, node: Node, l, r):  # 查询[l,r]区间的最小值
        if r < node.si or node.ei < l:
            return float('inf')
        if l <= node.si <= node.ei <= r:
            return node.valMin
        self.pushDown(node)
        ans = float('inf')
        if l <= node.left.ei:
            ans = min(ans, self.queryMin(node.left, l, r))
        if node.right.si <= r:
            ans = min(ans, self.queryMin(node.right, l, r))
        return ans


class StockPrice:

    def __init__(self):
        self.st = SegmentTree(1, 10**9)
        self.time = 0

    def update(self, timestamp: int, price: int) -> None:
        self.time = max(self.time, timestamp)
        self.st.update(self.st.root, timestamp, timestamp, price)

    def current(self) -> int:
        return self.st.queryMax(self.st.root, self.time, self.time)

    def maximum(self) -> int:
        return self.st.queryMax(self.st.root, 1, self.time)

    def minimum(self) -> int:
        return self.st.queryMin(self.st.root, 1, self.time)


# Your StockPrice object will be instantiated and called as such:
# obj = StockPrice()
# obj.update(timestamp,price)
# param_2 = obj.current()
# param_3 = obj.maximum()
# param_4 = obj.minimum()
```

```python
from sortedcontainers import SortedList


class StockPrice:

    def __init__(self):
        self.price = SortedList()
        self.time = 0
        self.time2price = {}

    def update(self, timestamp: int, price: int) -> None:
        self.price.discard(self.time2price.get(timestamp, 0))
        self.price.add(price)
        self.time = max(self.time, timestamp)
        self.time2price[timestamp] = price

    def current(self) -> int:
        return self.time2price[self.time]

    def maximum(self) -> int:
        return self.price[-1]

    def minimum(self) -> int:
        return self.price[0]


# Your StockPrice object will be instantiated and called as such:
# obj = StockPrice()
# obj.update(timestamp,price)
# param_2 = obj.current()
# param_3 = obj.maximum()
# param_4 = obj.minimum()
```
