```python
class Solution:

    def kthSmallest(self, mat: List[List[int]], k: int) -> int:
        m, n = len(mat), len(mat[0])

        def kSmallestPairs(a, b):  # lc373 两个数组最小的k对数组
            # 多路归并
            m, n = len(a), len(b)
            ans = []
            pq = [(a[i] + b[0], i, 0) for i in range(min(m, k))]
            while pq and len(ans) < k:
                s, i, j = heappop(pq)
                ans.append(s)
                j += 1
                if j < n:
                    heappush(pq, (a[i] + b[j], i, j))
            return ans

        a = mat[0][:k]
        for b in mat[1:]:
            a = kSmallestPairs(a, b)
        return a[-1]
```

```python
class Solution:

    def kthSmallest(self, mat: List[List[int]], k: int) -> int:
        rows, cols = len(mat), len(mat[0])

        def dfs(r, s, m):  # 和 <= m的元素数组数量
            nonlocal cnt
            if s > m or r == rows: return False
            # 选mat[r][0]
            if dfs(r + 1, s, m): return True
            # 不选mat[r][0]
            for c in range(1, cols):
                ns = s - mat[r][0] + mat[r][c]
                if ns > m: break
                cnt += 1
                if cnt >= k: return True
                if dfs(r + 1, ns, m): return True

        mn, mx = sum(row[0] for row in mat), sum(row[-1] for row in mat)
        l, r = mn, mx
        while l <= r:
            m = l + r >> 1
            cnt = 1  # 全选第一列
            if dfs(0, mn, m):
                r = m - 1
            else:
                l = m + 1
        return l
```

```python
class Solution:

    def kthSmallest(self, mat: List[List[int]], k: int) -> int:
        rows, cols = len(mat), len(mat[0])
        mn = sum(mat[i][0] for i in range(rows))
        mx = sum(mat[i][-1] for i in range(rows))

        def dfs(r, s, target):  # 和<=target 的数组数  target越大 数量越多
            nonlocal cnt
            if s > target or r == rows or cnt >= k:
                return
            dfs(r + 1, s, target)
            for i in range(1, cols):
                ns = s - mat[r][0] + mat[r][i]
                if ns > target:
                    break
                cnt += 1
                dfs(r + 1, ns, target)

        l, r = mn, mx
        while l <= r:
            m = l + r >> 1
            cnt = 1
            dfs(0, mn, m)
            if cnt >= k:
                r = m - 1
            else:
                l = m + 1
        return l
```
