```python
from sortedcontainers import SortedList


class Solution:

    def subarraysWithMoreZerosThanOnes(self, nums: List[int]) -> int:
        n = len(nums)
        vals = [0] * (4 * n)
        nums = (x * 2 - 1 for x in nums)
        # 统计区间和为正的区间数
        s = 0
        sl = SortedList([0])
        ans = 0
        for x in nums:
            s += x
            # s - s1 > 0 -> s1 < s
            r = sl.bisect_left(s) - 1
            ans += r + 1
            ans %= (10**9 + 7)
            sl.add(s)
        return ans
```
