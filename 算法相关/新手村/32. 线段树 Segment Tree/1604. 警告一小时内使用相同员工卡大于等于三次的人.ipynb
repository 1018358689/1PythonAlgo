```python
class SegTree:

    def __init__(self, n):  # [0,n - 1]
        self.vals = [0] * (4 * n)
        self.build(1, 0, n - 1)

    def build(self, o, l, r):  # 离散化后build
        if l == r:
            self.vals[o] = 1
            return
        m = l + r >> 1
        self.build(o * 2, l, m)
        self.build(o * 2 + 1, m + 1, r)
        self.vals[o] = self.vals[o * 2] + self.vals[o * 2 + 1]

    def query(self, o, l, r, L, R):  # 查询[L,R]
        if L <= l and r <= R:
            return self.vals[o]
        m = l + r >> 1
        ans = 0
        if m >= L:
            ans += self.query(o * 2, l, m, L, R)
        if m + 1 <= R:
            ans += self.query(o * 2 + 1, m + 1, r, L, R)
        return ans


class Solution:

    def alertNames(self, keyName: List[str], keyTime: List[str]) -> List[str]:
        ha = defaultdict(list)
        for k, t in zip(keyName, keyTime):
            ha[k].append(int(t[:2]) * 60 + int(t[-2:]))
        ans = []
        for name in sorted(ha.keys()):
            times = sorted(ha[name])
            n = len(times)
            sgt = SegTree(n)
            for i in range(n):
                l = bisect_right(times, times[i] + 60) - 1
                if sgt.query(1, 0, n - 1, i, l) >= 3:
                    ans.append(name)
                    break
        return ans
```

```python
class BinTree:

    def __init__(self, n):  # [0,n - 1]
        self.n = n
        self.vals = [0] * (n + 1)
        for x in range(1, n + 1):
            self.update(x)

    def update(self, x):
        while x <= self.n:
            self.vals[x] += 1
            x += x & -x

    def _query(self, lim):
        ans = 0
        while lim > 0:
            ans += self.vals[lim]
            lim -= lim & -lim
        return ans

    def query(self, L, R):
        return self._query(R) - self._query(L - 1)


class Solution:

    def alertNames(self, keyName: List[str], keyTime: List[str]) -> List[str]:
        ha = defaultdict(list)
        for k, t in zip(keyName, keyTime):
            ha[k].append(int(t[:2]) * 60 + int(t[-2:]))
        ans = []
        for name in sorted(ha.keys()):
            times = sorted(ha[name])
            n = len(times)
            bint = BinTree(n)
            for i in range(1, n):  # 偏移1位
                l = bisect_right(times, times[i - 1] + 60) - 1 + 1
                if bint.query(i, l) >= 3:
                    ans.append(name)
                    break
        return ans
```

```python
class Solution:

    def alertNames(self, keyName: List[str], keyTime: List[str]) -> List[str]:
        ha = defaultdict(list)
        for k, t in zip(keyName, keyTime):
            ha[k].append(int(t[:2]) * 60 + int(t[-2:]))
        ans = []
        for name in sorted(ha.keys()):
            times = sorted(ha[name])
            for x, y in zip(times, times[2:]):
                if y - x <= 60:
                    ans.append(name)
                    break
        return ans
```

```python
class Solution:

    def alertNames(self, keyName: List[str], keyTime: List[str]) -> List[str]:
        ha = defaultdict(list)
        for k, t in zip(keyName, keyTime):
            ha[k].append(t)
        ans = []
        for name in sorted(ha.keys()):
            times = sorted(ha[name])
            for x, y in zip(times, times[2:]):
                xh = x[:2]
                yh = y[:2]
                if xh == yh or (int(yh) == int(xh) + 1 and y[-2:] <= x[-2:]):
                    ans.append(name)
                    break
        return ans
```
