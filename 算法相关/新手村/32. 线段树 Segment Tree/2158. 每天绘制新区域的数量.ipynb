```python
# 线段树模板
# https://leetcode.cn/problems/amount-of-new-area-painted-each-day/solutions/1266646/pythonxian-duan-shu-chao-shi-bian-yuan-j-epaa/
```

```python
class Solution:

    def amountPainted(self, paint: List[List[int]]) -> List[int]:
        s = defaultdict(int)  # 区间和
        lazy = defaultdict(lambda: None)  # 待办未下发

        def push_down(o, l, r):
            if lazy[o] is None: return
            m = l + r >> 1
            s[o * 2] = (m - l + 1) * lazy[o]
            s[o * 2 + 1] = (r - m) * lazy[o]
            lazy[o * 2] = lazy[o * 2 + 1] = lazy[o]
            lazy[o] = None

        def update(o, l, r, ul, ur, val):  # [ul,ur]设置val
            if ul <= l and r <= ur:
                s[o] = (r - l + 1) * val
                lazy[o] = val
                return
            push_down(o, l, r)
            m = l + r >> 1
            if m >= ul:
                update(o * 2, l, m, ul, ur, val)
            if m + 1 <= ur:
                update(o * 2 + 1, m + 1, r, ul, ur, val)
            s[o] = s[o * 2] + s[o * 2 + 1]

        ans = []
        pre = 0
        for x, y in paint:
            update(1, 0, 50000, x, y - 1, 1)
            ans.append(s[1] - pre)
            pre = s[1]
        return ans
```

```python
class Solution:

    def amountPainted(self, paint: List[List[int]]) -> List[int]:
        n = 5 * 10**4
        s = [0] * (4 * n)  # 区间和
        lazy = [None] * (4 * n)  # 待办未下发

        def push_down(o, l, r):
            if lazy[o] is None: return
            m = l + r >> 1
            s[o * 2] = (m - l + 1) * lazy[o]
            s[o * 2 + 1] = (r - m) * lazy[o]
            lazy[o * 2] = lazy[o * 2 + 1] = lazy[o]
            lazy[o] = None

        def update(o, l, r, ul, ur, val):  # [ul,ur]设置val
            if r < ul or l > ur: return
            if ul <= l and r <= ur:
                s[o] = (r - l + 1) * val
                lazy[o] = val
                return
            push_down(o, l, r)
            m = l + r >> 1
            update(o * 2, l, m, ul, ur, val)
            update(o * 2 + 1, m + 1, r, ul, ur, val)
            s[o] = s[o * 2] + s[o * 2 + 1]

        ans = []
        pre = 0
        for x, y in paint:
            update(1, 0, 50000, x, y - 1, 1)
            ans.append(s[1] - pre)
            pre = s[1]
        return ans
```

```python
class Solution:
    def amountPainted(self, paint: List[List[int]]) -> List[int]:
        n = 5 * 10 ** 4 + 1
        roots = list(range(n))
        def find(x):
            if roots[x] != x:
                roots[x] = find(roots[x])
            return roots[x]
        def union(x, y):
            rx, ry = find(x), find(y)
            if rx != ry:
                roots[rx] = ry # rx -> ry
        ans = []
        for x, y in paint:
            x = find(x)
            y = find(y)
            cnt = 0
            while x < y:
                union(x, x + 1)
                x = find(x + 1)
                cnt += 1
            ans.append(cnt)
        return ans
```
