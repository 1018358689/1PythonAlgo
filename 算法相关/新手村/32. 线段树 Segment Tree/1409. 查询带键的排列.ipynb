```python
class Solution:

    def processQueries(self, queries: List[int], mm: int) -> List[int]:
        # 所谓坐标 即 该数前面数字个数
        # [0] * m + queries[->1]
        nn = len(queries)
        n = nn + mm
        vals = [0] * (4 * n)
        ha = {x: x + nn - 1 for x in range(1, mm + 1)}

        def build(o, l, r):
            if l == r:
                if l >= nn:
                    vals[o] = 1
                return
            m = l + r >> 1
            build(o * 2, l, m)
            build(o * 2 + 1, m + 1, r)
            vals[o] = vals[o * 2] + vals[o * 2 + 1]

        def update(o, l, r, idx, val):  # 单点设置
            if l == r:
                vals[o] = val
                return
            m = l + r >> 1
            if m >= idx:
                update(o * 2, l, m, idx, val)
            if m + 1 <= idx:
                update(o * 2 + 1, m + 1, r, idx, val)
            vals[o] = vals[o * 2] + vals[o * 2 + 1]

        def query(o, l, r, L, R):  # [L,R]取件查询
            if L <= l and r <= R:
                return vals[o]
            m = l + r >> 1
            ans = 0
            if m >= L:
                ans += query(o * 2, l, m, L, R)
            if m + 1 <= R:
                ans += query(o * 2 + 1, m + 1, r, L, R)
            return ans

        build(1, 0, n - 1)
        ans = []
        for i, x in enumerate(queries):
            pos = ha[x]
            ans.append(query(1, 0, n - 1, 0, pos - 1))
            update(1, 0, n - 1, pos, 0)
            update(1, 0, n - 1, nn - i - 1, 1)
            ha[x] = nn - i - 1
        return ans
```
