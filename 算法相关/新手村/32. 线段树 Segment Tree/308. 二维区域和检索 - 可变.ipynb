```python
class NumMatrix:

    def __init__(self, matrix: List[List[int]]):
        self.matrix = matrix
        self.rows, self.cols = len(matrix), len(matrix[0])
        self.n = self.rows * self.cols
        self.vals = [0] * (4 * self.n)
        self.build(1, 0, self.n - 1)

    def t(self, r, c):
        return r * self.cols + c

    def build(self, o, l, r):
        if l == r:
            row, col = divmod(l, self.cols)
            self.vals[o] = self.matrix[row][col]
            return
        m = l + r >> 1
        self.build(o * 2, l, m)
        self.build(o * 2 + 1, m + 1, r)
        self.vals[o] = self.vals[o * 2] + self.vals[o * 2 + 1]

    def _update(self, o, l, r, idx, val):  # idx = val
        if l == r:
            self.vals[o] = val
            return
        m = l + r >> 1
        if m >= idx:
            self._update(o * 2, l, m, idx, val)
        if m + 1 <= idx:
            self._update(o * 2 + 1, m + 1, r, idx, val)
        self.vals[o] = self.vals[o * 2] + self.vals[o * 2 + 1]

    def query(self, o, l, r, L, R):  # 查询[L,R]
        if L <= l and r <= R:
            return self.vals[o]
        m = l + r >> 1
        ans = 0
        if m >= L:
            ans += self.query(o * 2, l, m, L, R)
        if m + 1 <= R:
            ans += self.query(o * 2 + 1, m + 1, r, L, R)
        return ans

    def update(self, row: int, col: int, val: int) -> None:
        self._update(1, 0, self.n - 1, self.t(row, col), val)
        self.matrix[row][col] = val

    def sumRegion(self, row1: int, col1: int, row2: int, col2: int) -> int:
        ans = 0
        for r in range(row1, row2 + 1):
            id1 = self.t(r, col1)
            id2 = self.t(r, col2)
            ans += self.query(1, 0, self.n - 1, id1, id2)
        return ans


# Your NumMatrix object will be instantiated and called as such:
# obj = NumMatrix(matrix)
# obj.update(row,col,val)
# param_2 = obj.sumRegion(row1,col1,row2,col2)
```
