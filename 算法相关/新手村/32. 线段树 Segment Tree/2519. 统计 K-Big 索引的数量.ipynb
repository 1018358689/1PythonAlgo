```python
from sortedcontainers import SortedList
class Solution:
    def kBigIndices(self, nums: List[int], k: int) -> int:
        sl1 = SortedList()
        sl2 = SortedList(nums)
        ans = 0
        for x in nums:
            r = sl1.bisect_left(x) - 1 # <
            sl1.add(x)
            sl2.discard(x)
            if r + 1 < k: continue
            r = sl2.bisect_left(x) - 1 # <
            if r + 1 < k: continue
            ans += 1
        return ans
```

```python
# 超时
class SegmentTree:
    def __init__(self):
        self.s = defaultdict(int) # 区间和

    def update(self, o, l, r, idx, val): # 单点增加
        if l == r:
            self.s[o] += val
            return 
        m = l + r >> 1
        if m >= idx:
            self.update(o * 2, l, m, idx, val)
        else:
            self.update(o * 2 + 1, m + 1, r, idx, val)
        self.s[o] = self.s[o * 2] + self.s[o * 2 + 1]

    def query(self, o, l, r, ql, qr): #[ql, qr] 区间查询
        if ql <= l and r <= qr:
            return self.s[o]
        ans = 0
        m = l + r >> 1
        if m >= ql:
            ans += self.query(o * 2, l, m, ql, qr)
        if m + 1 <= qr:
            ans += self.query(o * 2 + 1, m + 1, r, ql, qr)
        return ans

class Solution:
    def kBigIndices(self, nums: List[int], k: int) -> int:
        n = len(nums)
        st1 = SegmentTree()
        st2 = SegmentTree()
        for x in nums:
            st2.update(1, 0, n, x, 1)
        ans = 0
        for x in nums:
            c1 = st1.query(1, 0, n, 1, x - 1)
            st1.update(1, 0, n, x, 1)
            st2.update(1, 0, n, x, -1)
            if c1 < k: continue
            c2 = st2.query(1, 0, n, 1, x - 1)
            if c2 < k: continue
            ans += 1
        return ans
```
