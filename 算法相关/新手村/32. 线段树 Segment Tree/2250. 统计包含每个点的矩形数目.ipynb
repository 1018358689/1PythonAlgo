```python
class Solution:

    def countRectangles(self, rectangles: List[List[int]], points: List[List[int]]) -> List[int]:
        n = len(points)
        m = len(rectangles)
        rectangles.sort(key=lambda x: -x[0])  # 倒序
        points = sorted(((x, y, i) for i, (x, y) in enumerate(points)), key=lambda x: -x[0])
        ans = [0] * n
        vals = defaultdict(int)  # 区间和

        def update(o, l, r, idx):  # += 1
            if l == r:
                vals[o] += 1
                return
            m = l + r >> 1
            if m >= idx:
                update(o * 2, l, m, idx)
            else:
                update(o * 2 + 1, m + 1, r, idx)
            vals[o] = vals[o * 2] + vals[o * 2 + 1]

        def query(o, l, r, lim):  # 查询[lim,]
            if lim <= l:
                return vals[o]
            m = l + r >> 1
            ans = query(o * 2 + 1, m + 1, r, lim)
            if m >= lim:
                ans += query(o * 2, l, m, lim)
            return ans

        j = 0
        for x, y, i in points:
            while j < m and rectangles[j][0] >= x:  # 固定x 把所有>=x的rectangles的y加入sl
                update(1, 1, 100, rectangles[j][1])
                j += 1
            ans[i] = query(1, 1, 100, y)  # = 所有>=y的个数 （x已经都满足）
        return ans
```

```python
from sortedcontainers import SortedList
class Solution:
    def countRectangles(self, rectangles: List[List[int]], points: List[List[int]]) -> List[int]:
        n = len(points)
        m = len(rectangles)
        rectangles.sort(key = lambda x:-x[0]) # 倒序
        points = sorted(((x, y, i) for i, (x, y) in enumerate(points)), key = lambda x:-x[0])
        ans = [0] * n
        sl = SortedList()
        j = 0
        for x, y, i in points:
            while j < m and rectangles[j][0] >= x: # 固定x 把所有>=x的rectangles的y加入sl
                sl.add(rectangles[j][1])
                j += 1
            ans[i] = len(sl) - sl.bisect_left(y) # = 所有>=y的个数 （x已经都满足）
        return ans
```
