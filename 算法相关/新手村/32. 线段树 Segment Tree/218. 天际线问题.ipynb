```python
# 超时
from typing import List


class Node:

    def __init__(self, si, ei) -> None:
        self.si = si
        self.ei = ei
        self.mi = si + ei >> 1
        self.left = None
        self.right = None
        self.val = 0  # 区间的最大值
        self.lazy = 0  # 设置值但未下发


class SegmentTree:

    def __init__(self, si, ei) -> None:
        self.root = Node(si, ei)

    def pushdown(self, node: Node):
        if not node.left: node.left = Node(node.si, node.mi)
        if not node.right: node.right = Node(node.mi + 1, node.ei)
        if not node.lazy:
            return
        node.left.val = max(node.left.val, node.lazy)
        node.right.val = max(node.right.val, node.lazy)
        node.left.lazy = max(node.left.lazy, node.lazy)
        node.right.lazy = max(node.right.lazy, node.lazy)
        node.lazy = 0

    def update(self, node: Node, l, r, val):  # [l,r]区间值设为val 与原来值比较
        if node.ei < l or node.si > r:
            return
        if l <= node.si and node.ei <= r:
            node.val = max(node.val, val)
            node.lazy = max(node.lazy, val)
            return
        self.pushdown(node)
        if l <= node.mi:
            self.update(node.left, l, r, val)
        if node.mi + 1 <= r:
            self.update(node.right, l, r, val)
        node.val = max(node.left.val, node.right.val)
        # node.val = node.left.val + node.right.val

    def query(self, node: Node, l, r):  # [l,r]区间最大值
        if node.ei < l or node.si > r:
            return 0
        if l <= node.si and node.ei <= r:
            return node.val
        self.pushdown(node)
        ans = 0
        if l <= node.mi:
            ans = max(ans, self.query(node.left, l, r))
        if node.mi + 1 <= r:
            ans = max(ans, self.query(node.right, l, r))
        return ans


class Solution:

    def getSkyline(self, buildings: List[List[int]]) -> List[List[int]]:
        left, right = buildings[0][0], buildings[-1][-2]
        st = SegmentTree(left, right)
        for l, r, v in buildings:
            st.update(st.root, l, r - 1, v)
        ans = []
        pre = 0
        for i in range(left, right + 1):
            cur = st.query(st.root, i, i)
            if cur != pre:
                ans.append([i, cur])
                pre = cur
        return ans


s = Solution()
s.getSkyline([[2, 9, 10], [3, 7, 15], [5, 12, 12], [15, 20, 10], [19, 24, 8]])
```
