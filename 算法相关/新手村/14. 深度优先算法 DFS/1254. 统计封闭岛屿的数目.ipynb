```python
class Solution:

    def closedIsland(self, grid: List[List[int]]) -> int:
        # 先外后内
        m, n = len(grid), len(grid[0])

        def dfs(i, j):
            grid[i][j] = 1  # 入x
            for ni, nj in (i + 1, j), (i - 1, j), (i, j + 1), (i, j - 1):
                if m > ni >= 0 <= nj < n and grid[ni][nj] == 0:
                    dfs(ni, nj)

        # 1. 把边界0进行向内填充
        for i in range(m):
            step = 1 if i in {0, m - 1} else n - 1
            for j in range(0, n, step):
                if grid[i][j] == 0:
                    dfs(i, j)
        # 2. 剩下0都能构成封闭岛
        ans = 0
        for i in range(1, m - 1):
            for j in range(1, n - 1):
                if grid[i][j] == 0:
                    ans += 1
                    dfs(i, j)
        return ans
```

```python
class Solution:

    def closedIsland(self, grid: List[List[int]]) -> int:
        # 出界标记
        m, n = len(grid), len(grid[0])

        def dfs(i, j):
            # 入x
            grid[i][j] = 1
            if i in {0, m - 1} or j in {0, n - 1}:
                nonlocal closed
                closed = False
            for ni, nj in (i + 1, j), (i - 1, j), (i, j + 1), (i, j - 1):
                if m > ni >= 0 <= nj < n and grid[ni][nj] == 0:
                    dfs(ni, nj)

        ans = 0
        for i in range(1, m - 1):
            for j in range(1, n - 1):
                if grid[i][j] == 0:
                    closed = True
                    dfs(i, j)
                    ans += closed
        return ans
```

```python
class Solution:

    def closedIsland(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid), len(grid[0])
        n = rows * cols
        roots = list(range(n + 1))
        count = n + 1  # count - 1

        def find(x):
            if roots[x] != x:
                roots[x] = find(roots[x])
            return roots[x]

        def union(x, y):
            rx, ry = find(x), find(y)
            if rx != ry:
                roots[rx] = ry
                nonlocal count
                count -= 1

        for r in range(rows):
            for c in range(cols):
                if grid[r][c] or r in {0, rows - 1} or c in {0, cols - 1}:
                    union(r * cols + c, n)
                if not grid[r][c]:
                    for nr, nc in (r + 1, c), (r, c + 1):
                        if nr < rows and nc < cols and not grid[nr][nc]:
                            union(r * cols + c, nr * cols + nc)
        return count - 1
```
