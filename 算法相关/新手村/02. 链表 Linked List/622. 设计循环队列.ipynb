```python
class Node:

    def __init__(self, val) -> None:
        self.val = val
        self.pre = None
        self.next = None


class DoubleList:

    def __init__(self) -> None:
        self.head = Node(-1)
        self.end = Node(-1)
        self.head.next = self.end
        self.end.pre = self.head

    def addAfter(self, node: Node, preNode: Node):
        nxt = preNode.next
        node.next = nxt
        node.pre = preNode
        preNode.next = node
        nxt.pre = node

    def delNode(self, node: Node):
        pre, nxt = node.pre, node.next
        pre.next = nxt
        nxt.pre = pre


class MyCircularQueue:

    def __init__(self, k: int):
        self.k = k
        self.dl = DoubleList()

    def enQueue(self, value: int) -> bool:
        if self.isFull():
            return False
        self.dl.addAfter(Node(value), self.dl.end.pre)
        self.k -= 1
        return True

    def deQueue(self) -> bool:
        if self.isEmpty():
            return False
        self.dl.delNode(self.dl.head.next)
        self.k += 1
        return True

    def Front(self) -> int:
        if self.isEmpty():
            return -1
        return self.dl.head.next.val

    def Rear(self) -> int:
        if self.isEmpty():
            return -1
        return self.dl.end.pre.val

    def isEmpty(self) -> bool:
        return self.dl.head.next == self.dl.end

    def isFull(self) -> bool:
        return not self.k


# Your MyCircularQueue object will be instantiated and called as such:
# obj = MyCircularQueue(k)
# param_1 = obj.enQueue(value)
# param_2 = obj.deQueue()
# param_3 = obj.Front()
# param_4 = obj.Rear()
# param_5 = obj.isEmpty()
# param_6 = obj.isFull()
```
