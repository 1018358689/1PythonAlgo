```python
from collections import Counter
from typing import List


class Solution:

    def groupStrings(self, words: List[str]) -> List[int]:
        # 1.转换到二进制集合（因为每个单词不含重复元素）
        def tran(s):
            ans = 0
            for ch in s:
                ans |= 1 << (ord(ch) - ord('a'))
            return ans

        words = [tran(word) for word in words]
        n = len(words)
        # 2.建立并查集
        roots = {word: word for word in words}
        sz = Counter(words)  # 单词间可能重复
        sets = set(words)

        def find(x):
            if roots[x] != x:
                roots[x] = find(roots[x])
            return roots[x]

        def union(x, y):
            if y not in sz:
                roots[y] = y
                sz[y] = 0
            rx, ry = find(x), find(y)
            if rx != ry:
                if sz[rx] > sz[ry]: rx, ry = ry, rx
                roots[rx] = ry
                sz[ry] += sz[rx]

        # 3.枚举合并
        for x in sets:
            for i in range(26):
                y = x | (1 << i)  # 只管添加 （删除会在被添加那个较长的单词进行）
                if y in sz:  # 存在则union
                    union(x, y)
                if x & (1 << i):  # 该位是1 进行替换 成 *
                    #比如abc中存在a，则制造*bc，（这与bc并不相同，所以将第27位置1以示区别）
                    #那么abc和*bc合并，同理之后某次还会有zbc与*bc合并，如此便间接将abc与zbc合并
                    y = (x - (1 << i)) | (1 << 26)
                    union(x, y)
        # 4.找集合数和大小
        chils = set()
        mx = 0
        for word in sets:
            fa = find(word)
            chils.add(fa)
            if sz[fa] > mx:
                mx = sz[fa]
        return [len(chils), mx]
```

```python
# 超时
class Solution:
    def groupStrings(self, words: List[str]) -> List[int]:
        def tran(s):
            ans = 0
            for ch in s:
                ans |= 1 << (ord(ch) - ord('a'))
            return ans
        words = [tran(word) for word in words]
        n = len(words)
        def check(s1, s2):
            x = s1 & s2
            d1 = (s1 - x).bit_count()
            d2 = (s2 - x).bit_count()
            return (d1 == 0 and d2 == 1) or (d1 == 1 and d2 == 0) or (d1 == 1 and d2 == 1)
        roots = {word:word for word in words}
        sz = Counter(words)
        def find(x):
            if roots[x] != x:
                roots[x] = find(roots[x])
            return roots[x]
        def union(x, y):
            rx, ry = find(x), find(y)
            if rx != ry:
                if sz[rx] > sz[ry]: rx, ry = ry, rx
                roots[rx] = ry
                sz[ry] += sz[rx]
        for i in range(n):
            for j in range(i + 1, n):
                x, y = words[i], words[j]
                # print(x, y, check(x, y))
                if check(x, y):
                    union(x, y)
        chils = set()
        mx = 0
        for word in words:
            fa = find(word)
            chils.add(fa)
            if sz[fa] > mx:
                mx = sz[fa]
        return [len(chils), mx]
```
