```python
class Solution:

    def regionsBySlashes(self, grid: List[str]) -> int:
        n = 3 * len(grid)
        gd = [[0] * n for _ in range(n)]
        for r in range(n // 3):
            for c in range(n // 3):
                if grid[r][c] == '/':
                    # gd[r * 3][c * 3 + 2] = 1
                    # gd[r * 3 + 1][c * 3 + 1] = 1
                    # gd[r * 3 + 2][c * 3] = 1
                    for i in range(3):
                        gd[r * 3 + i][c * 3 + 2 - i] = 1
                if grid[r][c] == '\\':
                    for i in range(3):
                        gd[r * 3 + i][c * 3 + i] = 1

        def dfs(r, c):
            for nr, nc in (r + 1, c), (r - 1, c), (r, c + 1), (r, c - 1):
                if n > nr >= 0 <= nc < n and gd[nr][nc] == 0:
                    gd[nr][nc] = 1
                    dfs(nr, nc)

        ans = 0
        for r in range(n):
            for c in range(n):
                if gd[r][c] == 0:
                    ans += 1
                    dfs(r, c)
        return ans
```
