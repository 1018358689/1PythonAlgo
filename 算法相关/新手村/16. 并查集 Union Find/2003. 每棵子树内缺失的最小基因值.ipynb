```python
class Solution:

    def smallestMissingValueSubtree(self, parents: List[int], nums: List[int]) -> List[int]:
        n = len(nums)
        adj = defaultdict(list)
        for i, x in enumerate(parents):
            adj[x].append(i)
        # 找基因1
        node1 = -1
        for i, x in enumerate(nums):
            if x == 1:
                node1 = i
                break
        # 利用并查集快速跳过已经访问的
        roots = list(range(max(nums) + 2))

        def find(x):
            if roots[x] != x:
                roots[x] = find(roots[x])
            return roots[x]

        def union(x, y):
            rx, ry = find(x), find(y)
            if rx != ry:
                roots[rx] = ry  # rx -> ry

        # 遍历子树 并记录遍历过的基因值（题目说基因值不重复 故可以用及一致作为vis判断）
        def dfs(x):
            union(nums[x], nums[x] + 1)
            for y in adj[x]:
                if find(nums[y]) == nums[y]:
                    dfs(y)

        # 往上走确定每个父节点的最小缺失基因值 会不断变大
        ans = [1] * n
        mex = 2
        p = node1
        while p != -1:
            dfs(p)
            mex = find(mex)
            ans[p] = mex
            p = parents[p]
        return ans
```

```python
class Solution:
    def smallestMissingValueSubtree(self, parents: List[int], nums: List[int]) -> List[int]:
        n = len(nums)
        adj = defaultdict(list)
        for i, x in enumerate(parents):
            adj[x].append(i)
        # 找基因1
        node1 = -1
        for i, x in enumerate(nums):
            if x == 1:
                node1 = i
                break
        # 遍历子树 并记录遍历过的基因值（题目说基因值不重复 故可以用及一致作为vis判断）
        vis = [False] * (max(nums) + 2)
        def dfs(x):
            vis[nums[x]] = True
            for y in adj[x]:
                if not vis[nums[y]]:
                    dfs(y)
        # 往上走确定每个父节点的最小缺失基因值 会不断变大
        ans = [1] * n
        mex = 2
        p = node1
        while p != -1:
            dfs(p)
            while vis[mex]:
                mex += 1
            ans[p] = mex
            p = parents[p]
        return ans
```
