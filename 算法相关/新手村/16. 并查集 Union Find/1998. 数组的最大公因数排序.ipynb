```python
class Solution:

    def gcdSort(self, nums: List[int]) -> bool:

        def fac(num):  # 分解质因数
            x = num
            p = 2
            while p**2 <= x:
                if x % p == 0:
                    while x % p == 0:
                        x //= p
                    yield p
                p += 1
            if x > 1: yield x

        roots = {}

        def find(x):
            if roots[x] != x:
                roots[x] = find(roots[x])
            return roots[x]

        def union(x, y):
            if x not in roots:
                roots[x] = x
            if y not in roots:
                roots[y] = y
            rx, ry = find(x), find(y)
            if rx != ry:
                roots[rx] = ry

        for x in nums:
            for y in fac(x):
                union(x, y)
        for x, y in zip(nums, sorted(nums)):
            if x != y and find(x) != find(y):
                return False
        return True
```

```python
# 预处理数据范围内所有数字的质因子
N = 10**5
prime = [[] for _ in range(N+1)]
for i in range(2, N+1):
    p, x = 2, i
    while p*p <= x:
        if x%p == 0:
            prime[i].append(p)
            while x%p == 0:
                x //= p
        p += 1
    if x > 1:
        prime[i].append(x)

class Solution:
    def gcdSort(self, nums: List[int]) -> bool:
        n = len(nums)

        # 并查集
        fa = list(range(n))
        def find(x: int) -> int:
            if fa[x] != x:
                fa[x] = find(fa[x])
            return fa[x]
        def merge(x: int, y: int) -> int:
            x, y = find(x), find(y)
            if x == y:
                return y
            fa[x] = y
            return y
        
        # 将具有同一质因子的数放到一起
        dic = defaultdict(list)
        for i, num in enumerate(nums):
            for p in prime[num]:
                dic[p].append(i)
        
        # 合并同一集合里的数
        for v in dic.values():
            reduce(merge, v)
        
        # 最后判断
        arr = sorted(zip(nums, range(n)))
        # print(arr)
        # print(fa)
        for i, (_, j) in enumerate(arr):
            if find(i) != find(j):
                return False
        return True
```
