```python
class Solution:

    def trulyMostPopular(self, names: List[str], synonyms: List[str]) -> List[str]:
        sys = []
        for sy in synonyms:
            x, y = sy.split(',')
            x, y = x[1:], y[:-1]
            sys.append((x, y))

        sz = {}
        rnames = set()
        for name in names:
            na, d = name.split('(')
            rnames.add(na)
            sz[na] = int(d[:-1])
        for x, y in sys:
            if x not in rnames: sz[x] = 0
            if y not in rnames: sz[y] = 0
        roots = {na: na for na in sz}

        def find(x):
            if roots[x] != x:
                roots[x] = find(roots[x])
            return roots[x]

        def union(x, y):
            rx, ry = find(x), find(y)
            if rx != ry:
                if rx < ry:
                    roots[ry] = rx
                    sz[rx] += sz[ry]
                else:
                    roots[rx] = ry
                    sz[ry] += sz[rx]

        for x, y in sys:
            union(x, y)
        chils = defaultdict(int)
        for na in rnames:
            chils[find(na)] = sz[find(na)]
        return [f'{na}({d})' for na, d in chils.items()]
```
