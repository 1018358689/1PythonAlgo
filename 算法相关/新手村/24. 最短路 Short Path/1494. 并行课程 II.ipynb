```python
from functools import cache
from itertools import combinations
from typing import List


class Solution:

    def minNumberOfSemesters(self, n: int, relations: List[List[int]], k: int) -> int:
        pre = [0] * n  # 前驱
        for x, y in relations:
            pre[y - 1] |= 1 << (x - 1)
        target = (1 << n) - 1  # 目标全完成

        @cache
        def dfs(sta):  # 从state到target最少步数
            if sta == target:
                return 0
            # 根据现在状态接下来 可学 的课
            canstudy = [i for i in range(n) if pre[i] | sta == sta and sta >> i & 1 == 0]
            m = len(canstudy)
            mn = (m + k - 1) // k  # 下限
            ans = n
            for ys in combinations(canstudy, min(m, k)):
                nsta = sta
                for y in ys:
                    nsta |= 1 << y
                ans = min(ans, dfs(nsta) + 1)
                if ans == mn: break
            return ans

        return dfs(0)
```

```python
from collections import deque
from typing import List
from itertools import combinations


class Solution:

    def minNumberOfSemesters(self, n: int, relations: List[List[int]], k: int) -> int:
        pre = [0] * n  # 前驱
        for x, y in relations:
            pre[y - 1] |= 1 << (x - 1)
        target = (1 << n) - 1  # 目标全完成
        step = 0
        q = deque([0])
        vis = [0] * (1 << n)  # 2 ** n 每个位置0或1
        vis[0] = 1
        while q:
            for _ in range(len(q)):
                sta = q.popleft()
                canstudy = [i for i in range(n) if pre[i] | sta == sta and sta >> i & 1 == 0]
                for ys in combinations(canstudy, min(len(canstudy), k)):
                    nsta = sta
                    for y in ys:
                        nsta |= 1 << y
                    if nsta == target: return step + 1
                    if vis[nsta]: continue
                    vis[nsta] = 1
                    q.append(nsta)
            step += 1
```

```python
from collections import defaultdict
from heapq import heappop, heappush
from math import inf
from typing import List
from itertools import combinations


class Solution:

    def minNumberOfSemesters(self, n: int, relations: List[List[int]], k: int) -> int:
        target = (1 << n) - 1
        pre = [0] * n  # 前驱
        for x, y in relations:
            pre[y - 1] |= (1 << (x - 1))
        pq = [(0, 0)]  # cost(step),state
        costs = defaultdict(lambda: inf)
        costs[0] = 0
        best = set()
        while pq:
            cost, state = heappop(pq)
            if state == target: return cost
            if state in best: continue
            best.add(state)
            can_study = []
            # 学过的 没法学（前驱不全） 就跳过
            for i in range(n):
                if (1 << i) & state: continue
                if pre[i] & state != pre[i]: continue  # pre[i]是子集代表前驱全
                can_study.append(i)
            if len(can_study) <= k:
                nstate = state
                for i in can_study:
                    nstate |= (1 << i)
                if cost + 1 < costs[nstate]:
                    costs[nstate] = cost + 1
                    heappush(pq, (cost + 1, nstate))
            else:
                for j in combinations(can_study, k):
                    nstate = state
                    for i in j:
                        nstate |= (1 << i)
                    if cost + 1 < costs[nstate]:
                        costs[nstate] = cost + 1
                        heappush(pq, (cost + 1, nstate))
```
