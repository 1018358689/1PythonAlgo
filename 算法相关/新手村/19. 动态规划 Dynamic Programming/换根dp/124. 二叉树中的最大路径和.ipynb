```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:

    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        ans = -inf

        def dfs(o):  # o为根的单边最大路径和 0表示无贡献
            if not o: return 0
            l_len = dfs(o.left)
            r_len = dfs(o.right)
            nonlocal ans
            ans = max(ans, l_len + r_len + o.val)
            return max(0, l_len + o.val, r_len + o.val)

        dfs(root)
        return ans
```
