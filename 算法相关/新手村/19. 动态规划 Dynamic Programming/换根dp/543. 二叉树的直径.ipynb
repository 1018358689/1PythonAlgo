```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:

    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        down1 = defaultdict(lambda: -1)
        down2 = defaultdict(lambda: -1)
        nr = {}
        up = defaultdict(lambda: -1)

        def dfs0(o):  # 返回最大深度
            if not o:
                return -1
            for y in o.left, o.right:
                mx = dfs0(y) + 1
                if mx > down1[o]:
                    down1[o], down2[o] = mx, down1[o]
                    nr[o] = y
                elif mx > down2[o]:
                    down2[o] = mx
            return down1[o]

        def dfs1(o):  # 用于更新向上最大高度
            if not o: return
            for y in o.left, o.right:
                if nr[o] == y:
                    up[y] = max(up[o], down2[o]) + 1
                else:
                    up[y] = max(up[o], down1[o]) + 1
                dfs1(y)

        dfs0(root)
        dfs1(root)
        ans = 0

        def dfs(o):
            if not o: return
            nonlocal ans
            ans = max(ans, down1[o] + up[o], down1[o] + down2[o])
            dfs(o.left)
            dfs(o.right)

        dfs(root)
        return ans
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:

    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        ans = 0

        def dfs(o):  # 以o为根的最大深度
            nonlocal ans
            if not o: return -1
            sub1 = dfs(o.left)
            sub2 = dfs(o.right)
            ans = max(ans, sub1 + sub2 + 2)
            return max(sub1, sub2) + 1

        dfs(root)
        return ans
```
