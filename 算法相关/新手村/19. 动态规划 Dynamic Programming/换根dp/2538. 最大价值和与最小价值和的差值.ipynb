```python
from typing import List


class Solution:
    def maxOutput(self, n: int, edges: List[List[int]], price: List[int]) -> int:
        # 和最大的一条路径 再减去该路径的 叶子端点的最小值
        # 最大直径题
        adj = [[] for _ in range(n)]
        for x, y in edges:
            adj[x].append(y)
            adj[y].append(x)

        ans = 0

        def dfs(x, fa):  # 子树中x为根 带叶子最大费用 不带叶子最大费用
            nonlocal ans
            mx1, mx2 = price[x], 0
            for y in adj[x]:
                if y != fa:
                    submx1, submx2 = dfs(y, x)
                    ans = max(ans, mx1 + submx2, mx2 + submx1)
                    mx1 = max(mx1, submx1 + price[x])
                    mx2 = max(mx2, submx2 + price[x])
            return mx1, mx2

        dfs(0, -1)
        return ans
```

```python
from math import inf


class Solution:
    def maxOutput(self, n: int, edges: List[List[int]], price: List[int]) -> int:
        # 换根dp
        adj = [[] for _ in range(n)]
        for x, y in edges:
            adj[x].append(y)
            adj[y].append(x)
        mx = [0] * n  # 向下路径最大值 (先以0为根)

        def dfs0(x, fa):  # 返回向下路径最大值
            ans = 0
            for y in adj[x]:
                if y != fa:
                    sub = dfs0(y, x)
                    ans = max(ans, sub)
            mx[x] = ans + price[x]
            return mx[x]

        def change(x, y1):  # x换到y1为根
            mx[x] = price[x]
            for y2 in adj[x]:
                if y2 != y1:
                    mx[x] = max(mx[x], mx[y2] + price[x])
            mx[y1] = max(mx[y1], mx[x] + price[y1])

        ans = -inf

        def dfs(x, fa):
            nonlocal ans
            ans = max(ans, mx[x] - price[x])
            for y in adj[x]:
                if y != fa:
                    change(x, y)
                    dfs(y, x)
                    change(y, x)
        dfs0(0, -1)
        dfs(0, -1)
        return ans
```
