```python
class Solution:

    def getMaxGridHappiness(self, m: int, n: int, intr: int, extr: int) -> int:
        # 从上往下 从左往右 一个一个填 计算当前位置 及 上 左 对整体答案贡献值
        # 当前位置上和左 就是 前n个 和 前1个
        # 每个位置 012 用长度n的三进制表示 当前位置 前面n个位置的放置状态 太麻烦了直接元祖表示好了
        score = [0, 120, 40]  # 代表分数
        cost = [[0, 0, 0], [0, -60, -10], [0, -10, 40]]  # 前面放置和当前放置的贡献度
        p = pow(3, n - 1)  # 最高位

        @lru_cache(None)
        def dfs(pos, intr, extr, state):
            if pos == m * n: return 0
            if intr == 0 and extr == 0: return 0
            up = state // p  # 最高位放置状态
            left = 0 if pos % n == 0 else state % 3  # 最低位放置状态 如果是第一列则强制0
            ans = 0
            for i in range(3):  # 枚举当前放置
                if (i == 1 and intr == 0) or (i == 2 and extr == 0):
                    continue
                nstate = state % p * 3 + i  # (state - up * p) * 3 + i # 去掉最高位
                a = score[i]
                b = dfs(pos + 1, intr - (i == 1), extr - (i == 2), nstate)
                c = cost[up][i] + cost[left][i]
                ans = max(ans, a + b + c)
            return ans

        return dfs(0, intr, extr, 0)
```

```python
class Solution:

    def getMaxGridHappiness(self, m: int, n: int, int: int, ext: int) -> int:
        # 从上往下 从左往右 一个一个填 计算当前位置 及 上 左 对整体答案贡献值
        # 当前位置上和左 就是 前n个 和 前1个
        # 每个位置 012 用长度n的三进制表示 当前位置 前面n个位置的放置状态 太麻烦了直接元祖表示好了
        @lru_cache(None)
        def dfs(pos, int, ext, state):
            if pos == m * n: return 0
            i, j = divmod(pos, n)
            state = list(state)
            ans = dfs(pos + 1, int, ext, tuple(state[1:] + [0]))  # 不放
            if int:  # 放内向
                diff = 120
                if state[0] == 1:  # 上面
                    diff -= 60
                elif state[0] == 2:
                    diff -= 10
                if j:  # 不是开头位置 可以计算左边
                    if state[-1] == 1:
                        diff -= 60
                    elif state[-1] == 2:
                        diff -= 10
                ans = max(ans, dfs(pos + 1, int - 1, ext, tuple(state[1:] + [1])) + diff)
            if ext:  # 放外向
                diff = 40
                if state[0] == 1:  # 上面
                    diff -= 10
                elif state[0] == 2:
                    diff += 40
                if j:  # 不是开头位置 可以计算左边
                    if state[-1] == 1:
                        diff -= 10
                    elif state[-1] == 2:
                        diff += 40
                ans = max(ans, dfs(pos + 1, int, ext - 1, tuple(state[1:] + [2])) + diff)
            return ans

        return dfs(0, int, ext, tuple([0] * n))
```
