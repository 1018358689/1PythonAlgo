```python
# yield 生成全排列 减少不必要内存开销
class Solution:

    def maxCompatibilitySum(self, students: List[List[int]], mentors: List[List[int]]) -> int:
        # 学生找导师 反过来也可以
        t = lambda lst: sum(x << i for i, x in enumerate(lst))
        a = [t(lst) for lst in students]
        b = [t(lst) for lst in mentors]
        m, n = len(students), len(students[0])  # 学生数和题目数
        cal = lambda x, y: n - (x ^ y).bit_count()

        # 无脑全排列
        def dfs():
            nonlocal vis
            # 入x
            if vis == u:
                yield path
            for j in range(m):  # vis是入y前的vis状态
                if vis >> j & 1 == 0:  # 判断入y
                    # 入y
                    vis |= 1 << j
                    path.append(a[j])
                    yield from dfs()
                    # 离y  y->x
                    vis ^= 1 << j
                    path.pop()

        vis = 0
        u = (1 << m) - 1
        path = []
        # 取最大
        ans = 0
        for st in dfs():
            ans = max(ans, sum(cal(x, y) for x, y in zip(st, b)))
        return ans
```
