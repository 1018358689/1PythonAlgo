```python
canfill = [0] * 15  # 每个位置可以填入哪些数 -1处理
for i in range(15):
    for j in range(15):
        if (i + 1) % (j + 1) == 0 or (j + 1) % (i + 1) == 0:
            canfill[i] |= 1 << j


class Solution:

    def countArrangement(self, n: int) -> int:
        u = (1 << n) - 1

        @lru_cache(None)
        def dfs(sta):  # [0:n]范围构造 已选位置 方案数
            i = sta.bit_count()
            if i == n: return 1
            ans = 0
            cf = (u ^ sta) & canfill[i]
            while cf:
                lb = cf & -cf
                ans += dfs(sta | lb)
                cf ^= lb
            return ans

        return dfs(0)
```
