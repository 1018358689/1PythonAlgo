```python
class Solution:

    def minimumTimeRequired(self, jobs: List[int], k: int) -> int:
        n = len(jobs)
        ans = inf
        ts = [0] * k  # 每个人的工作时间

        def dfs(i, mx):  # 逐个遍历每个工作 及当前最大工作时间
            nonlocal ans
            if mx > ans: return
            if i == n:
                ans = mx if mx < ans else ans
                return
            for j in range(k):  # 枚举给哪个人
                if j > 0 and ts[j] == ts[j - 1]:  # 枚举给相同的人 给一个情况就行 其他跳过
                    continue
                ts[j] += jobs[i]
                dfs(i + 1, mx if mx > ts[j] else ts[j])
                ts[j] -= jobs[i]

        dfs(0, 0)
        return ans
```
