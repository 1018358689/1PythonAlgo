```python
from functools import lru_cache
from typing import List


class Solution:

    def bestTeamScore(self, scores: List[int], ages: List[int]) -> int:
        sa = sorted(zip(scores, ages))
        n = len(sa)

        @lru_cache(None)
        def dfs(i):  # 以sa[i]结尾的最长递增子序列
            ans = sa[i][0]
            for j in range(i):
                if sa[j][1] <= sa[i][1]:
                    ans = max(ans, dfs(j) + sa[i][0])
            return ans

        return max(dfs(i) for i in range(n))
```

```python
from typing import List


class Solution:

    def bestTeamScore(self, scores: List[int], ages: List[int]) -> int:
        sa = sorted(zip(scores, ages))
        n = len(sa)

        f = [0] * n
        for i, (s, a) in enumerate(sa):
            f[i] = s
            for j in range(i):
                if sa[j][1] <= a:
                    f[i] = max(f[i], f[j] + s)
        return max(f)
```
