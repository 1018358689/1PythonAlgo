```python
from typing import List


class Solution:

    def countBits(self, n: int) -> List[int]:
        mem = [None] * (n + 1)
        mem[0] = 0

        def dfs(i):  # 数字i的1的个数
            if mem[i] is None:
                mem[i] = dfs(i - 1) + 1 if i & 1 else dfs(i >> 1)
            return mem[i]

        for i in range(n + 1):
            mem[i] = dfs(i)
        return mem


s = Solution()
s.countBits(101)
```
