```python
from typing import List
from collections import defaultdict, deque


class Solution:

    def magnificentSets(self, n: int, edges: List[List[int]]) -> int:
        fa = list(range(n + 1))
        sz = [1] * (n + 1)

        def find(x):
            if fa[x] != x:
                fa[x] = find(fa[x])
            return fa[x]

        def union(x, y):
            fx, fy = find(x), find(y)
            if fx == fy: return
            if sz[fx] > sz[fy]: fx, fy = fy, fx
            sz[fy] += sz[fx]
            fa[fx] = fy

        def get_group():
            group = defaultdict(list)
            for x in range(1, n + 1):
                group[find(x)].append(x)
            return group

        adj = [[] for _ in range(n + 1)]
        for x, y in edges:
            adj[x].append(y)
            adj[y].append(x)
            union(x, y)
        group = get_group()

        def bfs(x):  # 以x为根的最大深度
            q = deque([x])
            vis = {x}
            d = 0
            while q:
                for _ in range(len(q)):
                    x = q.popleft()
                    for y in adj[x]:
                        if y not in vis:
                            vis.add(y)
                            q.append((y))
                d += 1
            return d

        colors = [0] * (n + 1)

        def dfs(x, c):  # 检查是否二分图
            colors[x] = c
            for y in adj[x]:
                if colors[y] == c or colors[y] == 0 and not dfs(y, -c):
                    return False
            return True

        ans = 0
        for g in group.values():
            tmp = 0
            for x in g:
                if colors[x] == 0 and not dfs(x, 1):
                    return -1
                tmp = max(tmp, bfs(x))

            ans += tmp
        return ans


s = Solution()
s.magnificentSets(6, [[1, 2], [1, 4], [1, 5], [2, 6], [2, 3], [4, 6]])
```
