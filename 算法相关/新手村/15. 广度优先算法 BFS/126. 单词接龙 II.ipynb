```python
from collections import deque
from typing import List


class Solution:

    def findLadders(self, beginWord: str, endWord: str, wordList: List[str]) -> List[List[str]]:
        n = len(beginWord)
        abcd = [chr(x) for x in range(ord('a'), ord('z') + 1)]

        def adj(word):
            for i in range(n):
                for x in abcd:
                    yield word[:i] + x + word[i + 1:]

        wordList = set(wordList)
        if endWord not in wordList:
            return []
        q = deque([(beginWord, [beginWord])])
        vis = {beginWord}
        ans = []
        while q:
            nvis = set()  # 下一层重复访问不算
            for _ in range(len(q)):
                x, path = q.popleft()
                for y in adj(x):
                    if y in wordList and y not in vis:
                        if y == endWord:
                            ans.append(path + [y])
                        nvis |= {y}
                        q.append((y, path + [y]))
            if ans:
                break
            vis |= nvis
        return ans
```

```python
from typing import List
from collections import deque


class Node:
    def __init__(self, parent=None, val=-1) -> None:
        self.parent = parent
        self.val = val


class Solution:
    def neis(self, word: str):
        for i, n in enumerate(word):
            for j in range(ord('a'), ord('z') + 1):
                w = chr(j)
                if w != n:
                    yield word[:i] + w + word[i + 1:]

    def path(self, node: Node):
        ans = []
        while node:
            ans.append(node.val)
            node = node.parent
        return ans[::-1]

    def findLadders(self, beginWord: str, endWord: str, wordList: List[str]) -> List[List[str]]:
        wordList = set(wordList)
        if endWord not in wordList:
            return []
        q = deque([(beginWord, Node(val=beginWord))])
        vis = {beginWord}
        ans = []
        while q:
            neivis = set()
            for _ in range(len(q)):
                cur, node = q.popleft()
                for nei in self.neis(cur):
                    if nei not in vis and nei in wordList:
                        neinode = Node(node, nei)
                        if nei == endWord:
                            ans.append(self.path(neinode))
                        neivis.add(nei)
                        q.append((nei, neinode))
            vis |= neivis
            if ans:
                break
        return ans
```
