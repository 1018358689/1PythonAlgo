```python
from collections import defaultdict
from typing import List


class Solution:

    def numDistinctIslands(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid), len(grid[0])
        n = rows * cols
        roots = list(range(n + 1))

        def find(x):
            if roots[x] != x:
                roots[x] = find(roots[x])
            return roots[x]

        def union(x, y):
            rx, ry = find(x), find(y)
            if rx != ry:
                roots[rx] = ry

        dirs = (1, 0), (0, 1)
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == 0:
                    union(r * cols + c, n)
                else:
                    for dr, dc in dirs:
                        nr, nc = r + dr, c + dc
                        if rows > nr >= 0 <= nc < cols and grid[nr][nc] == 1:
                            union(r * cols + c, nr * cols + nc)
        chils = defaultdict(list)
        for node in range(n):
            fa = find(node)
            if fa != n:
                chils[fa].append(node)
        ans = set()
        for ch in chils.values():
            dr, dc = divmod(ch[0], cols)
            ans.add(tuple(map(lambda x: (x // cols - dr, x % cols - dc), ch)))
        return len(ans)
```

```python
from collections import deque
from typing import List


class Solution:

    def numDistinctIslands(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid), len(grid[0])
        ans = set()

        def bfs(start):
            q = deque([start])
            dr, dc = start
            points = []
            while q:
                r, c = q.popleft()
                points.append((r - dr, c - dc))
                for nr, nc in (r + 1, c), (r - 1, c), (r, c + 1), (r, c - 1):
                    if rows > nr >= 0 <= nc < cols and grid[nr][nc] == 1:
                        grid[nr][nc] = 0
                        q.append((nr, nc))
            return tuple(points)

        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == 1:
                    ans.add(bfs((r, c)))
        return len(ans)
```

```python
'''
计算图形相同，有两种情况：
可以翻转：计算图形中所有点两两之间的欧氏距离之和
只能平移：深度优先搜索记录所有遍历过的方向（只要尝试过，就记录路径）
''' 
from collections import deque
from typing import List


class Solution:

    def numDistinctIslands(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid), len(grid[0])
        dirs = (1, 0), (-1, 0), (0, 1), (0, -1)
        ans = set()

        def dfs(r, c):
            grid[r][c] = 0
            for d, (dr, dc) in enumerate(dirs):
                nr, nc = r + dr, c + dc
                nonlocal points
                points += str(d)
                if rows > nr >= 0 <= nc < cols and grid[nr][nc] == 1:
                    dfs(nr, nc)

        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == 1:
                    points = ''
                    dfs(r, c)
                    ans.add(points)
        return len(ans)
```
