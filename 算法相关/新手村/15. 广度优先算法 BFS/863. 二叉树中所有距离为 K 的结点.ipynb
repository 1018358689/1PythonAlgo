```python
from collections import defaultdict, deque
from typing import List


# Definition for a binary tree node.
class TreeNode:

    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None


class Solution:

    def distanceK(self, root: TreeNode, target: TreeNode, k: int) -> List[int]:
        if k == 0:
            return [target.val]
        adj = defaultdict(list)

        def dfs(x):
            if x.left:
                adj[x.val].append(x.left.val)
                adj[x.left.val].append(x.val)
                dfs(x.left)
            if x.right:
                adj[x.val].append(x.right.val)
                adj[x.right.val].append(x.val)
                dfs(x.right)

        dfs(root)
        q = deque([target.val])
        vis = {target.val}
        step = 0
        ans = []
        while q:
            tvis = set()
            for _ in range(len(q)):
                x = q.popleft()
                for y in adj[x]:
                    if y not in vis:
                        q.append(y)
                        tvis.add(y)
                        if step + 1 == k:
                            ans.append(y)

            if ans:
                return ans
            vis |= tvis
            step += 1
        return []
```
