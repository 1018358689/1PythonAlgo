```python
from typing import List
from collections import deque


class Solution:

    def cutOffTree(self, forest: List[List[int]]) -> int:
        rows, cols = len(forest), len(forest[0])
        if forest[0][0] == 0:
            return -1
        heights = []
        for r in range(rows):
            for c in range(cols):
                if forest[r][c]>1:
                    heights.append((forest[r][c], r, c))
        heights.sort(key=lambda x: x[0])
        def bfs(sr, sc, er, ec):
            q = deque([(sr, sc, 0)])
            vis = {(sr, sc)}
            while q:
                r, c, step = q.popleft()
                if (r, c) == (er, ec):
                    return step
                for nr, nc in ((r - 1, c), (r + 1, c), (r, c - 1), (r, c + 1)):
                    if 0 <= nr < rows and 0 <= nc < cols and (nr, nc) not in vis and forest[nr][nc] >= 1:
                        q.append((nr, nc, step + 1))
                        vis.add((nr, nc))
            return -1

        res = 0
        sr, sc = 0, 0
        for _, er, ec in heights:
            if (localStep := bfs(sr, sc, er, ec)) == -1:
                return -1
            sr, sc = er, ec
            res += localStep
        return res

s = Solution()
# s.cutOffTree([[1, 2, 3], [0, 0, 4], [7, 6, 5]])
# s.cutOffTree([[1, 2, 3], [0, 0, 0], [7, 6, 5]])
# s.cutOffTree([[2, 3, 4], [0, 0, 5], [8, 7, 6]])
# s.cutOffTree(
#     [
#     [54581641,64080174,24346381,69107959],
#     [86374198,61363882,68783324,79706116],
#     [  668150,92178815,89819108,94701471],
#     [83920491,22724204,46281641,47531096],
#     [89078499,18904913,25462145,60813308]
#     ]
# )
s.cutOffTree([[4,2,3],[0,0,1],[7,6,5]])
```

```python
from typing import List
import heapq


class Solution:

    def cutOffTree(self, forest: List[List[int]]) -> int:
        rows, cols = len(forest), len(forest[0])
        if forest[0][0] == 0:
            return -1
        # 因为要多源BFS 先把树高排排序
        heights = []
        for r in range(rows):
            for c in range(cols):
                if forest[r][c] > 1:
                    heights.append((forest[r][c], r, c))
        heights.sort(key=lambda x: x[0])

        # 启发函数
        def h(r, c, er, ec):
            return abs(er - r) + abs(ec - c)

        # 多源bfs
        def bfs(sr, sc, er, ec):
            pq = [(0 + h(sr, sc, er, ec), 0, sr, sc)]
            best = [[0] * cols for _ in range(rows)]
            costs = {(sr, sc): 0}
            while pq:
                _, step, r, c = heapq.heappop(pq)
                if (r, c) == (er, ec):
                    return step
                best[r][c] = 1
                for nr, nc in ((r - 1, c), (r + 1, c), (r, c - 1), (r, c + 1)):
                    if 0 <= nr < rows and 0 <= nc < cols and not best[nr][nc] and forest[nr][nc] >= 1:
                        if (nr, nc) not in costs or step + 1 < costs[(nr, nc)]:
                            costs[(nr, nc)] = step + 1
                            priority = step + 1 + h(nr, nc, er, ec)
                            heapq.heappush(pq, (priority, step + 1, nr, nc))
            return -1

        res = 0
        sr, sc = 0, 0
        for _, er, ec in heights:
            if (localStep := bfs(sr, sc, er, ec)) == -1:
                return -1
            sr, sc = er, ec
            res += localStep
        return res


s = Solution()
# s.cutOffTree([[1, 2, 3], [0, 0, 4], [7, 6, 5]])
# s.cutOffTree([[1, 2, 3], [0, 0, 0], [7, 6, 5]])
# s.cutOffTree([[2, 3, 4], [0, 0, 5], [8, 7, 6]])
# s.cutOffTree(
#     [
#     [54581641,64080174,24346381,69107959],
#     [86374198,61363882,68783324,79706116],
#     [  668150,92178815,89819108,94701471],
#     [83920491,22724204,46281641,47531096],
#     [89078499,18904913,25462145,60813308]
#     ]
# )
s.cutOffTree([[4, 2, 3], [0, 0, 1], [7, 6, 5]])
```

```python
from itertools import product
# 能到400ms 最快

class Solution:

    def cutOffTree(self, forest: List[List[int]]) -> int:
        trees = []
        m, n = len(forest), len(forest[0])
        for i, j in product(range(m), range(n)):  # product 避免两层for
            if forest[i][j] > 1:
                trees.append((forest[i][j], i, j))
        trees.sort()

        directions = [(-1, 0), (0, 1), (1, 0), (0, -1)]

        def bfs(start, end) -> int:
            x1, y1 = start
            x2, y2 = end

            tmp = [row[:] for row in forest]  # 用等大的数组来记录走过的位置
            q = deque([(x1, y1, 0)])
            while q:
                x, y, step = q.popleft()

                if x == x2 and y == y2:  # 找到了
                    return step

                tmp[x][y] = 0
                for dx, dy in directions:
                    i, j = x + dx, y + dy
                    if 0 <= i < m and 0 <= j < n and tmp[i][j]:
                        # 精髓在这里：蹲一个大佬的解释。。。
                        if (dx == 0 and dy == 1 and y < y2) \
                                or (dx == 0 and dy == -1 and y > y2) \
                                or (dx == 1 and dy == 0 and x < x2) \
                                or (dx == -1 and dy == 0 and x > x2):
                            q.appendleft((i, j, step + 1))

                        else:
                            q.append((i, j, step + 1))

            return 0

        ans = 0

        if trees[0][1:] != (0, 0):  # 判空
            trees = [(1, 0, 0)] + trees

        for start, end in pairwise(trees):
            step = bfs(start[1:], end[1:])
            if not step: return -1

            forest[end[1]][end[2]] = 1
            ans += step
        return ans
```
