```python
# 从空地出发超时
from collections import deque
from math import inf
from typing import List


class Solution:

    def shortestDistance(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid), len(grid[0])
        one = sum(grid[r][c] == 1 for r in range(rows) for c in range(cols))

        def bfs(start):
            q = deque([(*start, 0)])
            vis = {start}
            ans = 0
            cnt = 0
            while q:
                r, c, step = q.popleft()
                if grid[r][c] == 1:
                    cnt += 1
                    ans += step
                    continue
                if cnt == one:
                    break
                for nr, nc in (r + 1, c), (r - 1, c), (r, c + 1), (r, c - 1):
                    if rows > nr >= 0 <= nc < cols and (nr, nc) not in vis and grid[nr][nc] != 2:
                        vis.add((nr, nc))
                        q.append((nr, nc, step + 1))
            return ans if cnt == one else inf

        ans = inf
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == 0:
                    ans = min(ans, bfs((r, c)))
        return ans if ans != inf else -1
```
