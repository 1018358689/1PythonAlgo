```python
from typing import Optional
from collections import defaultdict, deque
from math import inf


# Definition for a binary tree node.
class TreeNode:

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:

    def treeQueries(self, root: Optional[TreeNode], queries: List[int]) -> List[int]:
        # 1.每个节点高度
        high = {}  # 节点为根的高度

        def dfs(o: TreeNode):
            if not o:
                return -1
            l, r = dfs(o.left), dfs(o.right)
            ans = max(l, r) + 1
            high[o.val] = ans
            return ans

        h = dfs(root)
        # 2.每一层只有最高的树被去掉时,才需要减去(最高树-次高树),而删除该层其他所有节点,都不会导致树高的下降.
        d = defaultdict(lambda: 0)  # 节点相对树高要减去的值
        q = deque([root])
        # bfs预处理
        while q:
            one, two = (-inf, 0), (-inf, 0)
            for _ in range(len(q)):
                cur = q.popleft()
                num = high[cur.val]
                if num > one[0]:
                    two = one
                    one = (num, cur.val)
                elif num > two[0]:
                    two = (num, cur.val)
                if cur.left: q.append(cur.left)
                if cur.right: q.append(cur.right)
            if two[1] == 0:  # 一层只有一个节点 减去子树
                d[one[1]] = high[one[1]] + 1
            else:  # 差值为 最高-次高
                d[one[1]] = high[one[1]] - high[two[1]]
        ans = []
        for x in queries:
            ans.append(h - d[x])
        return ans
```
