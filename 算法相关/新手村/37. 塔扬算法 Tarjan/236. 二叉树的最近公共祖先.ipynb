```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


class Solution:

    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if root in {None, p, q}:
            return root
        l, r = self.lowestCommonAncestor(root.left, p, q), self.lowestCommonAncestor(root.right, p, q)
        if l and r:
            return root
        return l if l else r
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


class Solution:

    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        N = 10**5
        mx = 10**9 + 1
        jok = TreeNode(mx)
        m = N.bit_length()
        pa = defaultdict(lambda: defaultdict(lambda: jok))
        dep = defaultdict(int)

        def dfs(x, fa):
            pa[x][0] = fa
            dep[x] = dep[fa] + 1
            for i in range(1, m):
                if (p := pa[x][i - 1]) == jok: break
                pa[x][i] = pa[p][i - 1]
            for y in x.left, x.right:
                if y:
                    dfs(y, x)

        dfs(root, TreeNode(mx))
        if dep[p] < dep[q]:  # 保证p在下面
            p, q = q, p
        d = dep[p] - dep[q]
        for i in range(d.bit_length()):
            if d >> i & 1:
                p = pa[p][i]
                if p == jok: break
        if p == q: return q
        for i in range(m - 1, -1, -1):
            if pa[p][i] != pa[q][i]:
                p = pa[p][i]
                q = pa[q][i]
        return pa[p][0]
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


class Solution:

    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        roots = {}
        colors = defaultdict(int)

        def find(x):
            if roots[x] != x:
                roots[x] = find(roots[x])
            return roots[x]

        query = {p: q, q: p}

        def tarjan(x):
            roots[x] = x
            colors[x] = 1
            for y in x.left, x.right:
                if y is None: continue
                sub = tarjan(y)
                if sub is not None: return sub
                roots[y] = x
            if x in query and colors[query[x]] == 2:
                return find(query[x])
            colors[x] = 2
            return None

        return tarjan(root)
```
