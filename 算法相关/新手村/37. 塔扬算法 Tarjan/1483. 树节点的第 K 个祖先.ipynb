```python
from math import ceil, log
from typing import List


class TreeAncestor:

    def __init__(self, n: int, parent: List[int]):
        # m = n.bit_length()  # 一条链 至多跳 2**m步 绰绰有余
        m = ceil(log(n, 2))
        adj = [[] for _ in range(n)]
        for x, p in enumerate(parent):
            if p == -1: continue
            adj[p].append(x)
        pa = [[-1] * m for _ in range(n)]  # pa[x][i]: x节点的第2**i个祖先节点

        def dfs(x, fa):
            pa[x][0] = fa
            for i in range(1, m):
                if (p := pa[x][i - 1]) == -1: break
                pa[x][i] = pa[p][i - 1]
            for y in adj[x]:
                dfs(y, x)

        dfs(0, -1)
        self.pa = pa

    def getKthAncestor(self, node: int, k: int) -> int:
        for i in range(k.bit_length()):
            if k >> i & 1:  # 第i位是否为1 是则跳过去
                node = self.pa[node][i]
                if node < 0: break
        return node

    def getKthAncestor2(self, node: int, k: int) -> int:
        while node != -1 and k:
            lb = k & -k
            node = self.pa[node][lb.bit_length() - 1]
            k ^= lb
        return node


# Your TreeAncestor object will be instantiated and called as such:
# obj = TreeAncestor(n, parent)
# param_1 = obj.getKthAncestor(node,k)
```

```python
class TreeAncestor:

    def __init__(self, n: int, parent: List[int]):
        pa = [[] for _ in range(n)]  # pa[x][i]: x节点的第2**i个祖先节点
        for i, x in enumerate(parent):
            pa[i].append(x)
        j = 0
        allneg = False  # 所有为 -1
        while not allneg:
            j += 1  # 当前操作的索引
            allneg = True
            for i in range(n):
                t = pa[p][j - 1] if (p := pa[i][j - 1]) != -1 else -1
                pa[i].append(t)
                if t != -1: allneg = False
        self.pa = pa
        self.m = j + 1  # 列宽

    def getKthAncestor(self, node: int, k: int) -> int:
        for i in range(k.bit_length()):
            if i > self.m - 1: return -1
            if k >> i & 1:  # 第i位是否为1 是则跳过去
                node = self.pa[node][i]
                if node < 0: break
        return node

    def getKthAncestor2(self, node: int, k: int) -> int:
        while node != -1 and k:
            lb = k & -k
            pos = lb.bit_length() - 1
            if pos > self.m - 1: return -1
            node = self.pa[node][lb.bit_length() - 1]
            k ^= lb
        return node


# Your TreeAncestor object will be instantiated and called as such:
# obj = TreeAncestor(n, parent)
# param_1 = obj.getKthAncestor(node,k)
```
