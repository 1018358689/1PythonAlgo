```python
class Node:

    def __init__(self) -> None:
        self.children = {}
        self.end = False


class Trie:

    def __init__(self) -> None:
        self.root = Node()

    def add(self, word):
        cur = self.root
        for w in word:
            if w not in cur.children:
                cur.children[w] = Node()
            cur = cur.children[w]
        cur.end = True

    def match(self, word):
        cur = self.root
        for w in word:
            if w not in cur.children:
                return False
            cur = cur.children[w]
            if cur.end:
                return True
        return False


from typing import List


class StreamChecker:

    def __init__(self, words: List[str]):
        self.trie = Trie()
        for word in words:
            self.trie.add(word[::-1])
        self.w = ''

    def query(self, letter: str) -> bool:
        self.w = letter + self.w
        return self.trie.match(self.w)


# Your StreamChecker object will be instantiated and called as such:
# obj = StreamChecker(words)
# param_1 = obj.query(letter)
```
