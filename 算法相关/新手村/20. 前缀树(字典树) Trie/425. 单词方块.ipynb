```python
class Node:

    def __init__(self):
        self.child = {}
        self.ws = set()


class Trie:

    def __init__(self):
        self.root = Node()

    def add(self, word):
        cur = self.root
        for w in word:
            if w not in cur.child:
                cur.child[w] = Node()
            cur = cur.child[w]
            cur.ws.add(word)

    def query(self, word):  # 是否是前缀
        cur = self.root
        for w in word:
            if w not in cur.child:
                return []
            cur = cur.child[w]
        return cur.ws


class Solution:

    def wordSquares(self, words: List[str]) -> List[List[str]]:
        n = len(words[0])
        ans = []
        trie = Trie()
        for word in words:
            trie.add(word)

        def dfs(i, path):  # 和当前要确定的i列
            if i == n:
                ans.append(path[:])
                return
            pre = ''.join(w[i] for w in path)  # 要找的前缀（i列累积）
            for w in trie.query(pre):
                dfs(i + 1, path + [w])

        for w in words:
            dfs(1, [w])
        return ans
```
