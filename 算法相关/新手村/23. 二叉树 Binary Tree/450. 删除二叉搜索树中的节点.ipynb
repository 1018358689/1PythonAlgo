```python
# Definition for a binary tree node.
class TreeNode:

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


from typing import Optional


class Solution:

    def getSuccessor(self, root):
        pre = root
        cur = root.right
        while cur.left:
            pre = cur
            cur = cur.left
        return pre, cur

    def deleteNode(self, root: Optional[TreeNode], key: int) -> Optional[TreeNode]:
        if not root:
            return None
        # 1. 搜索key值得节点 并记录父节点
        pre, cur = None, root
        while cur and cur.val != key:
            pre = cur
            if cur.val < key:
                cur = cur.right
            else:
                cur = cur.left
        if cur is None:
            return root
        # 2.1 key节点是叶子节点
        if cur.left is None and cur.right is None:
            if cur == root:
                root = None
            else:
                # 删除key节点 需要判断他是左节点还是右节点才能让父节点继承
                if pre.left == cur:
                    pre.left = None
                else:
                    pre.right = None
        # 2.2 key节点是 只有左节点或右节点
        elif cur.left and cur.right is None:
            if cur == root:
                root = cur.left
            else:
                # 删除key节点 需要判断他是左节点还是右节点才能让父节点继承
                if pre.left == cur:
                    pre.left = cur.left
                else:
                    pre.right = cur.left
        elif cur.right and cur.left is None:
            if cur == root:
                root = cur.right
            else:
                if pre.left == cur:
                    pre.left = cur.right
                else:
                    pre.right = cur.right
        # 2.3 key节点 左右节点都有
        else:
            # 寻找后继节点
            pre2, cur2 = self.getSuccessor(cur)
            cur.val = cur2.val
            # 删除该后继节点 把他的右节点继承过来
            if pre2.left == cur2:
                pre2.left = cur2.right
            else:
                pre2.right = cur2.right
        return root
```

```python
# Definition for a binary tree node.
class TreeNode:

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


from typing import Optional


class Solution:

    def getSuccessor(self, root):
        cur = root.right
        while cur.left:
            cur = cur.left
        return cur

    def deleteNode(self, root: Optional[TreeNode], key: int) -> Optional[TreeNode]:
        # basecase
        if not root:
            return None

        if key > root.val:  # key在右边
            root.right = self.deleteNode(root.right, key)
        elif key < root.val:  # key在左边
            root.left = self.deleteNode(root.left, key)
        else:  # 找到key
            if root.left is None and root.right is None:  # key是叶子节点
                root = None
            elif root.left and root.right is None:  # key是只有左或右子树
                root = root.left
            elif root.right and root.left is None:
                root = root.right
            else:  # key左右子树都有
                suc = self.getSuccessor(root)  # 获得后驱节点
                root.val = suc.val  # 改值 然后删除
                root.right = self.deleteNode(root.right, suc.val)
        return root
```
