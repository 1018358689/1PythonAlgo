```python
class LockingTree:

    def __init__(self, parent: List[int]):
        n = len(parent)
        self.adj = [[] for _ in range(n)]
        for y, x in enumerate(parent):
            if x != -1:
                self.adj[x].append(y)
        self.states = [0] * n  # 给索引上锁用户 0表示未上锁
        self.parent = parent

    def lock(self, num: int, user: int) -> bool:
        if self.states[num] == 0:
            self.states[num] = user
            return True
        return False

    def unlock(self, num: int, user: int) -> bool:
        if self.states[num] == user:
            self.states[num] = 0
            return True
        return False

    def upgrade(self, num: int, user: int) -> bool:
        if self.states[num] != 0: return False
        p = self.parent[num]
        while p != -1:
            if self.states[p] > 0:
                return False
            p = self.parent[p]
        locked = False

        def dfs(x):
            if self.states[x] > 0:
                self.states[x] = 0
                nonlocal locked
                locked = True
            for y in self.adj[x]:
                dfs(y)

        dfs(num)
        if locked: self.states[num] = user
        return locked


# Your LockingTree object will be instantiated and called as such:
# obj = LockingTree(parent)
# param_1 = obj.lock(num,user)
# param_2 = obj.unlock(num,user)
# param_3 = obj.upgrade(num,user)
```
