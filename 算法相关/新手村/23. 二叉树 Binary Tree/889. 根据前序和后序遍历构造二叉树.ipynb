```python
# Definition for a binary tree node.
from typing import List


class TreeNode:

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:

    def constructFromPrePost(self, preorder: List[int], postorder: List[int]) -> TreeNode:
        if not preorder:
            return None
        if len(preorder) == 1:
            return TreeNode(preorder[0])
        root = TreeNode(preorder[0])
        lt = preorder[1] # 把第二个值作为左子树根
        m = postorder.index(lt)
        l = self.constructFromPrePost(preorder[1:2 + m], postorder[:m + 1])
        r = self.constructFromPrePost(preorder[2 + m:], postorder[m + 1:len(postorder) - 1])
        root.left = l
        root.right = r
        return root
```

```python
# Definition for a binary tree node.
from typing import List


class TreeNode:

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:

    def constructFromPrePost(self, preorder: List[int], postorder: List[int]) -> TreeNode:
        ha = {}
        for i, n in enumerate(postorder):
            ha[n] = i

        def dfs(l1, r1, l2, r2):
            if l1 > r1:
                return None
            if l1 == r1:
                return TreeNode(preorder[l1])
            root = TreeNode(preorder[l1])
            lt = preorder[l1 + 1]
            m = ha[lt]
            l = dfs(l1 + 1, m - l2 + l1 + 1, l2, m)
            r = dfs(m - l2 + l1 + 2, r1, m + 1, r2 - 1)
            root.left = l
            root.right = r
            return root

        return dfs(0, len(preorder) - 1, 0, len(postorder) - 1)
```
