```python
class MyCircularDeque:

    def __init__(self, k: int):
        self.k = k
        self.q = [0] * k
        self.head = self.cnt = 0

    def insertFront(self, value: int) -> bool:
        if self.isFull():
            return False
        self.head -= 1
        self.cnt += 1
        self.q[self.head % self.k] = value
        return True

    def insertLast(self, value: int) -> bool:
        if self.isFull():
            return False
        self.cnt += 1
        self.q[(self.head + self.cnt - 1) % self.k] = value
        return True

    def deleteFront(self) -> bool:
        if self.isEmpty():
            return False
        self.head += 1
        self.cnt -= 1
        if self.isEmpty():
            self.head = 0
        return True

    def deleteLast(self) -> bool:
        if self.isEmpty():
            return False
        self.cnt -= 1
        if self.isEmpty():
            self.head = 0
        return True

    def getFront(self) -> int:
        if self.isEmpty():
            return -1
        return self.q[self.head % self.k]

    def getRear(self) -> int:
        if self.isEmpty():
            return -1
        return self.q[(self.head + self.cnt - 1) % self.k]

    def isEmpty(self) -> bool:
        return self.cnt == 0

    def isFull(self) -> bool:
        return self.cnt == self.k


# Your MyCircularDeque object will be instantiated and called as such:
# obj = MyCircularDeque(k)
# param_1 = obj.insertFront(value)
# param_2 = obj.insertLast(value)
# param_3 = obj.deleteFront()
# param_4 = obj.deleteLast()
# param_5 = obj.getFront()
# param_6 = obj.getRear()
# param_7 = obj.isEmpty()
# param_8 = obj.isFull()
```
