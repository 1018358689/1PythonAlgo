```python
N = 10**5
lpf = list(range(N + 1))  # 最小质因数 埃式筛法 lpf[x]==x表示质数
for x in range(2, int(N**.5) + 1):
    if lpf[x] == x:
        for y in range(x * x, N + 1, x):
            lpf[y] = x # 合数的最小质因数


class Solution:

    def canTraverseAllPairs(self, nums: List[int]) -> bool:
        n = len(nums)
        roots = list(range(n))
        count = n

        def find(x):
            if roots[x] != x:
                roots[x] = find(roots[x])
            return roots[x]

        def union(x, y):
            rx, ry = find(x), find(y)
            if rx != ry:
                roots[rx] = ry  # rx -> ry
                nonlocal count
                count -= 1

        adj = defaultdict(set)
        for i, x in enumerate(nums):
            while x > 1:
                p = lpf[x]
                adj[p].add(i) # 质因数对应数字索引 会重复p所以add去重i
                x //= p
        for vals in adj.values():
            for x, y in pairwise(vals):
                union(x, y)
        return count == 1
```
