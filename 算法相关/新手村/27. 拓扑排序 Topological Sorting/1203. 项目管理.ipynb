```python
class Solution:

    def sortItems(self, n: int, m: int, group: List[int], beforeItems: List[List[int]]) -> List[int]:
        # 组间 and 组内 toposort
        gs = defaultdict(set)  # 小组-项目
        for i, x in enumerate(group):
            if x == -1:
                group[i] = m
                m += 1
            gs[group[i]].add(i)
        adj1 = defaultdict(set)  # 组间
        adj2 = defaultdict(lambda: defaultdict(set))  # 组内
        for x, bs in enumerate(beforeItems):
            for y in bs:  # y <- x
                g1, g2 = group[x], group[y]  # 项目所属组 g2 <- g1
                if g1 == g2:
                    adj2[g1][x].add(y)
                else:
                    adj1[g1].add(g2)

        def topo(nodes, adj):
            colors = defaultdict(int)
            res = []

            # 0:未访问 1访问过 2安全节点
            def dfs(x):  # 检测绝对环
                if colors[x] == 1: return True
                if colors[x] == 2: return False
                colors[x] = 1
                for y in adj[x]:
                    if dfs(y): return True
                colors[x] = 2
                res.append(x)
                return False

            for x in nodes:
                if dfs(x): return None
            return res

        res = []
        group_topo = topo(list(range(m)), adj1)
        if group_topo is None: return []
        for g in group_topo:
            inner_topo = topo(gs[g], adj2[g])
            if inner_topo is None: return []
            res += inner_topo
        return res
```

```python
class Solution:

    def sortItems(self, n: int, m: int, group: List[int], beforeItems: List[List[int]]) -> List[int]:
        # 组间 and 组内 toposort
        gs = defaultdict(set)  # 小组-项目
        for i, x in enumerate(group):
            if x == -1:
                group[i] = m
                m += 1
            gs[group[i]].add(i)
        adj1 = defaultdict(set)  # 组间
        adj2 = defaultdict(lambda: defaultdict(set))  # 组内
        for x, bs in enumerate(beforeItems):
            for y in bs:  # y -> x
                g1, g2 = group[x], group[y]  # 项目所属组 g2 -> g1
                if g1 == g2:
                    adj2[g1][y].add(x)
                else:
                    adj1[g2].add(g1)

        def topo(nodes, adj):
            ins = defaultdict(int)
            for x in nodes:
                for y in adj[x]:
                    ins[y] += 1
            q = deque([x for x in nodes if ins[x] == 0])
            res, cnt = [], 0
            while q:
                x = q.popleft()
                res.append(x)
                cnt += 1
                for y in adj[x]:
                    ins[y] -= 1
                    if ins[y] == 0:
                        q.append(y)
            return res if cnt == len(nodes) else None

        res = []
        group_topo = topo(list(range(m)), adj1)
        if group_topo is None: return []
        for g in group_topo:
            inner_topo = topo(gs[g], adj2[g])
            if inner_topo is None: return []
            res += inner_topo
        return res
```
