```python
from collections import defaultdict, deque
from typing import List


class Solution:

    def sortItems(self, n: int, m: int, group: List[int], beforeItems: List[List[int]]) -> List[int]:
        # 1.补组号 顺便 确定组内成员
        g2item = defaultdict(list)
        for x in range(n):
            if group[x] == -1:
                group[x] = m
                m += 1
            g2item[group[x]].append(x)
        # 2.确定组间和组内邻接表
        adj_group = defaultdict(list)  # 组间
        adj_in = defaultdict(lambda: defaultdict(list))  # 组内
        for y, b in enumerate(beforeItems):
            for x in b:
                gx, gy = group[x], group[y]
                if gx == gy:
                    adj_in[gx][x].append(y)
                else:
                    adj_group[gx].append(gy)
        # 3.拓扑排序
        def bfs(nodes, adj):
            ins = defaultdict(int)
            for sons in adj.values():
                for y in sons:
                    ins[y] += 1
            q = deque([x for x in nodes if not ins[x]])
            ans = []
            cnt = 0
            while q:
                x = q.popleft()
                ans.append(x)
                cnt += 1
                for y in adj[x]:
                    ins[y] -= 1
                    if not ins[y]:
                        q.append(y)
            return ans if cnt == len(nodes) else []

        # 3.分别组间和组内拓扑排序
        gs = bfs(list(set(group)), adj_group)
        if not gs:
            return []
        ans = []
        for g in gs:
            tmp = bfs(g2item[g], adj_in[g])
            if not tmp:
                return []
            ans += tmp
        return ans
```
