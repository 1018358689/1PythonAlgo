```python
from collections import deque
from typing import List


class Solution:

    def collectTheCoins(self, coins: List[int], edges: List[List[int]]) -> int:
        n = len(coins)
        adj = [[] for _ in range(n)]
        indeg = [0] * n
        for x, y in edges:
            adj[x].append(y)
            adj[y].append(x)
            indeg[x] += 1
            indeg[y] += 1

        left_edges = n - 1
        # topo sort 去掉无金币子树
        q = deque([])
        for i, (d, c) in enumerate(zip(indeg, coins)):
            if d == 1 and c == 0:
                q.append(i)
        while q:
            x = q.popleft()
            left_edges -= 1
            for y in adj[x]:
                indeg[y] -= 1
                if indeg[y] == 1 and coins[y] == 0:
                    q.append(y)
        # 两次topo sort 去掉含金币叶子节点和 其（去掉后）叶子父节点
        for i, (d, c) in enumerate(zip(indeg, coins)):
            if d == 1 and c == 1:
                q.append(i)
        left_edges -= len(q)

        for x in q:
            for y in adj[x]:
                indeg[y] -= 1
                if indeg[y] == 1:
                    left_edges -= 1

        return max(left_edges * 2, 0)
```

```python
from collections import deque
from typing import List


class Solution:

    def collectTheCoins(self, coins: List[int], edges: List[List[int]]) -> int:
        n = len(coins)
        adj = [[] for _ in range(n)]
        indeg = [0] * n
        for x, y in edges:
            adj[x].append(y)
            adj[y].append(x)
            indeg[x] += 1
            indeg[y] += 1

        # topo sort 去掉无金币子树
        q = deque([])
        for i, (d, c) in enumerate(zip(indeg, coins)):
            if d == 1 and c == 0:
                q.append(i)
        while q:
            x = q.popleft()
            for y in adj[x]:
                indeg[y] -= 1
                if indeg[y] == 1 and coins[y] == 0:
                    q.append(y)

        # 记录入队时间 （以叶子金币为基准的高度）
        time = [-1] * n
        for i, (d, c) in enumerate(zip(indeg, coins)):
            if d == 1 and c == 1:
                q.append(i)
                time[i] = 0
        while q:
            x = q.popleft()
            for y in adj[x]:
                indeg[y] -= 1
                if indeg[y] == 1:
                    time[y] = time[x] + 1
                    q.append(y)
        return sum(time[x] >= 2 and time[y] >= 2 for x, y in edges)
```
