```python
# 剑指 Offer II 002. 二进制加法
class Solution:
    def addBinary(self, a: str, b: str) -> str:
        maxlgh = max(len(a), len(b))

        def get(s, i):  # 补位获取
            d = maxlgh - len(s)
            if i - d >= 0:
                return int(s[i - d])
            else:
                return 0

        ans = ''
        carry = 0
        for i in range(maxlgh - 1, -1, -1):
            n1, n2 = get(a, i), get(b, i)
            carry, tmp = divmod(n1 + n2 + carry, 2)
            ans += str(tmp)
        if carry == 1:
            ans += '1'
        return ans[::-1]


class Solution2:
    def addBinary(self, a: str, b: str) -> str:

        ans = ''
        carry = 0
        p1,p2 = len(a)-1,len(b)-1
        while p1>=0 or p2>=0:
            n1 = int(a[p1]) if p1>=0 else 0
            n2 = int(b[p2]) if p2>=0 else 0
            carry,tmp = divmod(n1+n2+carry,2)
            ans+=str(tmp)
            if p1>=0:
                p1-=1
            if p2>=0:
                p2-=1
        if carry:
            ans+=str(carry)
        return ans[::-1]
```

```python
# lc2. 两数相加
from typing import Optional


# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode()
        p1, p2, cur = l1, l2, dummy
        carry = 0
        while p1 or p2:
            n1 = p1.val if p1 else 0
            n2 = p2.val if p2 else 0
            carry, tmp = divmod(n1 + n2 + carry, 10)
            cur.next = ListNode(tmp)
            cur = cur.next
            if p1: p1 = p1.next
            if p2: p2 = p2.next
        if carry:
            cur.next = ListNode(carry)
        return dummy.next
```

```python
# lc989. 数组形式的整数加法
from typing import List


class Solution:
    def addToArrayForm(self, num: List[int], k: int) -> List[int]:
        ans = []
        carry = 0
        p1 = len(num) - 1
        while p1 >= 0 or k:
            n1 = num[p1] if p1 >= 0 else 0
            n2 = k % 10 if k else 0
            carry, tmp = divmod(n1 + n2 + carry, 10)
            ans.append(tmp)
            if p1 >= 0:
                p1 -= 1
            if k:
                k //= 10
        if carry:
            ans.append(carry)
        return ans[::-1]
```

```python
# lc66. 加一
class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        carry, tmp = divmod(digits[-1] + 1, 10)
        p = len(digits) - 2
        ans = [tmp]
        while p >= 0 and carry: # 当digits遍历完 或 碰到不进位 则退出
            carry, tmp = divmod(digits[p] + 1, 10)
            ans.append(tmp)
            if p >= 0:
                p -= 1
        # 看哪种形式退出
        if carry == 0:
            return digits[:p + 1] + ans[::-1]
        else:
            if carry:
                ans.append(carry)
            return ans[::-1]


s = Solution()
s.plusOne([9, 9, 9])
```

```python
# lc415. 字符串相加
class Solution:
    def addStrings(self, num1: str, num2: str) -> str:
        carry = 0
        ans = ''
        p1, p2 = len(num1) - 1, len(num2) - 1
        while p1 >= 0 or p2 >= 0:
            n1 = ord(num1[p1]) - ord('0') if p1 >= 0 else 0
            n2 = ord(num2[p2]) - ord('0') if p2 >= 0 else 0
            carry, tmp = divmod(n1 + n2 + carry, 10)
            ans += str(tmp)
            if p1 >= 0:
                p1 -= 1
            if p2 >= 0:
                p2 -= 1
        if carry:
            ans += '1'
        return ans[::-1]
```

```python
# lc445. 两数相加 II
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        def tran(root: ListNode):
            if not root or not root.next:
                return root
            tmp = tran(root.next)
            root.next.next = root
            root.next = None
            return tmp

        l1 = tran(l1)
        l2 = tran(l2)
        p1, p2 = l1, l2
        dummy = ListNode()
        cur = dummy
        carry = 0
        while p1 or p2:
            n1 = p1.val if p1 else 0
            n2 = p2.val if p2 else 0
            carry, tmp = divmod(n1 + n2 + carry, 10)
            cur.next = ListNode(tmp)
            cur = cur.next
            if p1:
                p1 = p1.next
            if p2:
                p2 = p2.next
        if carry:
            cur.next = ListNode(carry)
        return tran(dummy.next)


class Solution2:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        stack1, stack2 = [], []
        cur = l1
        while cur:
            stack1.append(cur.val)
            cur = cur.next
        cur = l2
        while cur:
            stack2.append(cur.val)
            cur = cur.next
        dummy = ListNode()
        carry = 0
        while stack1 or stack2:
            n1 = stack1.pop() if stack1 else 0
            n2 = stack2.pop() if stack2 else 0
            carry, tmp = divmod(n1 + n2 + carry, 10)
            #  自己画图理解
            cur = ListNode(tmp)
            cur.next = dummy.next
            dummy.next = cur
        if carry:
            cur = ListNode(carry)
            cur.next = dummy.next
            dummy.next = cur
        return dummy.next
```
