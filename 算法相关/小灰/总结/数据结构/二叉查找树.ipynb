```python
class TreeNode:

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class BinarySearchTree:

    def __init__(self, root: TreeNode):
        self.root = root

    def search(self, data: int):
        target = self.root
        while target and target.val != data:
            if data > target.val:
                target = target.right
            else:
                target = target.left
        if target is None:
            return None
        else:
            return target

    def insert(self, data: int):
        node = TreeNode(data)
        if self.root is None:
            self.root = node
            return True
        target = self.root
        while True:
            if data == target.val:
                return False
            elif data > target.val:
                if target.right is None:
                    target.right = node
                    return True
                target = target.right
            else:
                if target.left is None:
                    target.left = node
                    return True
                target = target.left

    def delete(self, data: int):
        target = self.root
        parentNode = TreeNode()
        while target and target.val != data:
            parentNode = target
            if data > target.val:
                target = target.right
            else:
                target = target.left
        if target is None:
            return False
        # 待删除节点没有子节点
        if target.left is None and target.right is None:
            if target == self.root:
                self.root = None
            elif parentNode.left == target:
                parentNode.left = None
            else:
                parentNode.right = None
        # 只有一个左孩子
        elif target.left and target.right is None:
            if target == self.root:
                self.root = target.left
            elif parentNode.left == target:
                parentNode.left = target.left
            else:
                parentNode.right = target.left
        # 只有一个右孩子
        elif target.left is None and target.right:
            if target == self.root:
                self.root = target.right
            elif parentNode.left == target:
                parentNode.left = target.right
            else:
                parentNode.right = target.right
        # 左右孩子都有
        else:
            closeParentNode = target
            closeNode = target.right  # 待删除节点的后继节点 即最靠近他且大于的他的值
            while closeNode.left:
                closeParentNode = closeNode
                closeNode = closeNode.left
            target.val = closeNode.val
            if closeNode.right is None:
                if closeParentNode.left == closeNode:
                    closeParentNode.left = None
                else:
                    closeParentNode.right = None
            else:
                if closeParentNode.left == closeNode:
                    closeParentNode.left = closeNode.right
                else:
                    closeParentNode.right = closeNode.right
            # if closeParentNode.left == closeNode:
            #     closeParentNode.left = closeNode.right
            # else:
            #     closeParentNode.right = closeNode.right
        return True


# ln = TreeNode(5)
# rn = TreeNode(13)
# root = TreeNode(9, ln, rn)
root = None
t = BinarySearchTree(root)
t.insert(6), t.root.val
```
