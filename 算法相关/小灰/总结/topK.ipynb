topK问题 列出<=num的数量 然后二分取左

```python
# lc719. 找出第 K 小的数对距离
from typing import List


class Solution:

    def smallestDistancePair(self, nums: List[int], k: int) -> int:
        nums.sort()

        def d2cnt(d):  # <=d的数对个数 非单调递增
            ans = 0
            l = 0
            for r in range(len(nums)):
                while nums[r] - nums[l] > d:
                    l += 1
                if l != r:
                    ans += r - l
            return ans

        l, r = 0, nums[-1] - nums[0]
        while l <= r:
            m = l + r >> 1
            if d2cnt(m) >= k:  # 如[1,1,6]对应数对距离[0,5,5]当m为0,1,2,3,4都为1但是要取最左边 因为只有最左边存在
                r = m - 1
            else:
                l = m + 1
        return l
```

```python
# lc668. 乘法表中第k小的数
class Solution:

    def findKthNumber(self, m: int, n: int, k: int) -> int:

        def num2cnt(num):  # <=num的数量 非单调递增
            ans = 0
            for r in range(1, m + 1):
                ans += min(n, num // r)
            return ans

        l, r = 1, m * n
        while l <= r:
            mid = l + r >> 1
            if num2cnt(mid) >= k:  # 比如<=5个数和<=4个数一样都是6 但是5不存在所以往左取4
                r = mid - 1
            else:
                l = mid + 1
        return l
```

```python
# lc378. 有序矩阵中第 K 小的元素
class Solution:

    def kthSmallest(self, matrix: List[List[int]], k: int) -> int:
        n = len(matrix)

        def num2cnt(num):  # <=num个数
            i, j = 0, n - 1
            # 指针从第一行最右边开始比较
            # 如果num大于其值就以为大于其左边所有 指针下移
            # 如果小于则 指针左移
            cnt = 0
            while i <= n-1 and j >= 0:
                if num >= matrix[i][j]:
                    cnt += j + 1
                    i += 1
                else:
                    j -= 1
            return cnt

        l, r = matrix[0][0], matrix[-1][-1]
        while l <= r:
            m = l + r >> 1
            if num2cnt(m) >= k:
                r = m - 1
            else:
                l = m + 1
        return l
```

```python
# lc215. 数组中的第K个最大元素
from typing import List
import heapq


class Solution:

    def findKthLargest(self, nums: List[int], k: int) -> int:
        pq = []
        for n in nums:
            heapq.heappush(pq, -n)
        for _ in range(k - 1):
            heapq.heappop(pq)
        return -pq[0]
```

```python
# lc215. 数组中的第K个最大元素
from typing import List
from random import randint


class Solution:

    def findKthLargest(self, nums: List[int], k: int) -> int:

        def getPivot(si, ei):
            pivot = randint(si, ei)
            nums[si], nums[pivot] = nums[pivot], nums[si]
            pivot = nums[si]
            mark = si
            for i in range(si + 1, ei + 1):
                if nums[i] < pivot:
                    mark += 1
                    nums[i], nums[mark] = nums[mark], nums[i]
            nums[mark], nums[si] = nums[si], nums[mark]
            return mark

        target = len(nums) - k

        def dc(si, ei):
            pivot = getPivot(si, ei)
            if pivot > target:
                return dc(si, pivot - 1)
            elif pivot < target:
                return dc(pivot + 1, ei)
            else:
                return nums[pivot]

        return dc(0, len(nums) - 1)
```

```python
# lc75. 颜色分类
class Solution:

    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        mark = 0
        for i in range(len(nums)):
            if nums[i] == 0:
                nums[mark], nums[i] = nums[i], nums[mark]
                mark += 1
        for i in range(mark, len(nums)):
            if nums[i] == 1:
                nums[mark], nums[i] = nums[i], nums[mark]
                mark += 1
```

```python
# lc324. 摆动排序 II
from typing import List
from random import randint


class Solution:

    def wiggleSort(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """

        def getPivot(si, ei):
            pivot = randint(si, ei)
            nums[si], nums[pivot] = nums[pivot], nums[si]
            mark = si
            for i in range(si + 1, ei + 1):
                if nums[i] < nums[si]:
                    mark += 1
                    nums[mark], nums[i] = nums[i], nums[mark]
            nums[si], nums[mark] = nums[mark], nums[si]
            return mark

        lgh = len(nums)
        mi = 0 + lgh - 1 >> 1

        # topK
        def dc(si, ei):
            pivot = getPivot(si, ei)
            if pivot > mi:
                dc(si, pivot - 1)
            elif pivot < mi:
                dc(pivot + 1, ei)
            else:
                return

        # 需要三路划分(荷兰旗) 否则像这个出错[1, 3, 2, 2, 3, 1]
        def three_way_partition(nums, val):
            # [0,p1) <val
            # [p1,p2) =val
            # [p3,lgh) >val
            p1, p2, p3 = 0, 0, len(nums)
            while p2 < p3:
                if nums[p2] < val:
                    nums[p1], nums[p2] = nums[p2], nums[p1]
                    p1 += 1
                    p2 += 1
                elif nums[p2] == val:
                    p2 += 1
                else:
                    p3 -= 1
                    nums[p3], nums[p2] = nums[p2], nums[p3]
                    # p2 += 1

        dc(0, lgh - 1)
        three_way_partition(nums, nums[mi])

        # print(nums)
        # nums[::2], nums[1::2] = nums[:mi + 1][::-1], nums[mi + 1:][::-1]
        l, r, copyNums = mi, lgh - 1, nums[:]
        for i in range(lgh):
            if i & 1 == 1:
                nums[i] = copyNums[r]
                r -= 1
            else:
                nums[i] = copyNums[l]
                l -= 1
        print(nums)


s = Solution()
s.wiggleSort([1, 3, 2, 2, 3, 1])
s.wiggleSort([1, 2, 2, 5, 2, 1, 5, 2, 1, 3, 4, 4, 1, 5, 2])
```
