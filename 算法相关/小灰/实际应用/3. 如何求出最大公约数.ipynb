```python
def getGreatestCommonDivisor(a, b):
    big, small = max(a, b), min(a, b)
    if big % small == 0:
        return small
    for n in range(small // 2, 0, -1):
        if big % n == 0 and small % n == 0:
            return n


getGreatestCommonDivisor(2, 3), getGreatestCommonDivisor(6, 9)
```

```python
# 辗转相除法
# 缺点:取模运算性能差
def getGreatestCommonDivisor(a, b):
    big, small = max(a, b), min(a, b)
    if (div := big % small) == 0:
        return small
    return getGreatestCommonDivisor(div, small)


getGreatestCommonDivisor(6, 9), getGreatestCommonDivisor(1, 10000)
```

```python
# 更相减损术
# 缺点:递归次数大 如求1,10000需要递归9999次


def getGreatestCommonDivisor(a, b):
    if a == b:
        return a
    big, small = max(a, b), min(a, b)
    return getGreatestCommonDivisor(big - small, small)


getGreatestCommonDivisor(6, 9), getGreatestCommonDivisor(1, 10000)
```

```python
# 辗转相除+更相减损术结合 Stein算法
def getGreatestCommonDivisor(a, b):
    if a == b:
        return a
    if (a & 1) == 0 and (b & 1) == 0:  # 都是偶数提取2
        return getGreatestCommonDivisor(a >> 1, b >> 1) << 1
    elif a & 1 == 0:  # 其中一个偶数 则排除2
        return getGreatestCommonDivisor(a >> 1, b)
    elif b & 1 == 0:
        return getGreatestCommonDivisor(a, b >> 1)
    else:  # 都是奇数则更相减损术
        big, small = max(a, b), min(a, b)
        return getGreatestCommonDivisor((big - small) >> 1, small)  # 奇数相减为偶数


getGreatestCommonDivisor(6, 9), getGreatestCommonDivisor(1, 10000)
```

```python
# 更相减损法,求最大公约数
# https://baike.baidu.com/item/%E6%9B%B4%E7%9B%B8%E5%87%8F%E6%8D%9F%E6%9C%AF/449183?fr=aladdin 
# 同样求1,10000需要递归9999次
def gcf(a, b):
    # 减少倍数
    reduction = 1
    while True:
        # 可半者半之
        if a % 2 == 0 and b % 2 == 0:
            a //= 2
            b //= 2
            reduction *= 2
        else:
            # 不可半者
            break
    # cnt = 0
    while True:
        # 副置分母、子之数，以少减多，更相减损，求其等也
        big = max(a, b)
        small = min(a, b)
        # cnt+=1
        # print(big,small,cnt)
        if small * 2 == big:
            return small * reduction
        else:
            a, b = small, big - small


# gcf(6, 9) 
gcf(1, 10000)
```
