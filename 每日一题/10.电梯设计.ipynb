```python
## 某高层住宅（楼层-2~25）共有2个电梯，请为该住宅设计一个电梯调度程序：
## 1.输入4个参数，参数1、2分别代表两个电梯当前位置，参数3代表电梯即将下行或的上行方向，参数4代表需要乘坐电梯的人群所在楼层。
## 2.输出1个结果（1或2），代表即将承担本次运输的电梯编号。
```

```python
class Elevator:
    def __init__(self, name, maxFloor, minFloor):
        self.name = name
        self.maxFloor = maxFloor
        self.minFloor = minFloor
        self.currentPos = 1

    def set_currentPos(self, newPos):
        if newPos > self.maxFloor or newPos < self.minFloor:
            raise Exception(
                f'超出电梯楼层范围，正确楼层应在{self.minFloor}至{self.maxFloor}层。')
        else:
            self.currentPos = newPos

    def get_relaPos(self, pos):
        if self.currentPos < 0 and pos > 0:
            relaPos = pos - (self.currentPos + 1)
        if self.currentPos > 0 and pos < 0:
            relaPos = (pos + 1) - self.currentPos
        else:
            relaPos = pos - self.currentPos
        return relaPos


e1 = Elevator('Elevator-1', 25, -2)
e2 = Elevator('Elevator-2', 25, -2)

e1.set_currentPos(2)
e2.set_currentPos(-2)

pressAt = -1

if abs(e1.get_relaPos(pressAt)) > abs(e2.get_relaPos(pressAt)):
    print('即将由{}号电梯承载本次运载任务。'.format(e2.name))
else:
    print('即将由{}号电梯承载本次运载任务。'.format(e1.name))
```

```python
from threading import Timer
import math


#class evevator
class Elevator(object):
    def __init__(self, name, maxFloor=25, minFloor=1):
        self.name = name
        self.maxFloor = maxFloor
        self.minFloor = minFloor
        self.pos = 1
        self.posLst = []
        self.stat = '空闲'

    def set_NewPos(self, newPos):
        if newPos > self.maxFloor or newPos < self.minFloor:
            raise Exception(
                "Err:out of range! The num must between %d and %d." %
                (self.minFloor, self.maxFloor))
        self.pos = newPos

    def get_relaPos(self, pos):  #获取pos楼层与本电梯当前的相对位置。注意：没有0层
        if self.stat == '运行中':
            return math.inf  #无穷大
        if self.pos < 0 and pos > 0:
            res = pos - self.pos - 1
        elif self.pos > 0 and pos < 0:
            res = pos - self.pos + 1
        else:
            res = pos - self.pos
        return res

    def set_posLst(self, pos):
        if self.pos < pos:
            self.posLst = list(range(self.pos, pos + 1))
        else:
            self.posLst = list(range(pos, self.pos + 1))
        try:
            self.posLst.remove(0)
        except:
            pass

        if pos < self.pos:  #下行方向
            self.posLst.reverse()

        return

    def run(self):
        if len(self.posLst) > 0:
            self.stat = '运行中'
            self.pos = self.posLst[0]
            self.posLst.remove(self.posLst[0])
            t = Timer(1, self.run)
            t.start()
        else:
            print("%s号电梯已到达指定位置%d层" % (self.name, self.pos))
            self.stat = '空闲'


e1 = Elevator("1", 25, -2)
e2 = Elevator("2", 25, -2)

while True:
    pressAt = int(input("请输入你所在的楼层："))

    obj = e1
    if e1.stat == '运行中' and e2.stat == '运行中':
        print('所有电梯都在运行中，请稍候！')
    else:
        if abs(e1.get_relaPos(pressAt)) > abs(
                e2.get_relaPos(pressAt)):  # 由距离近的电梯承担本次任务
            obj = e2

        print("即将由%s号电梯承担本次运载任务。" % (obj.name))
        obj.set_posLst(pressAt)

        print(obj.posLst)
        obj.run()
```
