今日题（1110）：兑换喝酒
描述：小区便利店正在促销，用 n个空酒瓶可以兑换一瓶新酒。你购入了m瓶酒。请你计算最多能喝到多少瓶酒。
举例：
输入：m=9，n=3
输出：13
 
输入：m=15，n=4
输出：19

```python
def f(m, n):
    cnt = m
    while m >= n:
        quo, rem = divmod(m, n)
        cnt += quo
        m = quo + rem
    return cnt


f(9, 3), f(15, 4), f(5, 5), f(9, 5)
```

```python
# 瓶酒分离，n瓶=1瓶酒=1瓶+1酒 -> n-1瓶=1酒
# m瓶=m//(n-1)酒 因为最后手上肯定有瓶 避免出现借瓶喝酒 -> m瓶=(m-1)//(n-1)酒
def f(m, n):
    return m + (m - 1) // (n - 1)

f(9, 3), f(15, 4), f(5, 5), f(9, 5)
```

```python
# 一瓶一瓶喝 达到n则对换一瓶酒
def f(m,n):
    l = list(range(1,m+1))
    for i in l:
        print(l)
        if i%n==0:
            m+=1
            l.append(m)
    return m
f(9, 3), f(15, 4), f(5, 5), f(9, 5)
```
