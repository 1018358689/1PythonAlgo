# 对迭代器、生成器做切片操作
# 已知函数fib()可以创建一个斐波那契生成器对象，对象可以生成1,1,2,3,5,8,13,21...无限长度的斐波那契数列。
# 现在需要对该生成器对象做切片操作，找出第1001至1005的五个数分别为多少？

```python
def fib():
    prev, curr = 0, 1
    while True:
        yield curr
        curr, prev = prev + curr, curr
f = fib()
print(f)
```

```python
from itertools import islice
for i in islice(f,1001,1006):
    print(i)
```

本来是直接拿去用fib()函数  功能就是 斐波那契数列 生成器。   
现在简单解释一下这个生成器。 这个涉及到容器、迭代器、生成器的概念。
容器是一种把多个元素组织在一起的数据结构，可以逐个迭代获取其中的元素。例如列表、集合、元组、字典，大多是容器都是可迭代对象。
迭代器是指实现了__iter__和__next__方法的对象都称为迭代器。迭代器是一个有状态的对象，只有在调用时才生成值，没有调用的时候就等待下一次调用。
生成器其实是一种特殊的迭代器，但是不需要像迭代器一样实现__iter__和__next__方法，只需要使用关键字yield就可以。生成器的优点是延迟计算，一次返回一个结果，这样非常适用于大数据量的计算。但是，使用生成器必须要注意的一点是：生成器只能遍历一次。

迭代器有个重要特性，就是可以记住迭代状态，以便于下次迭代时接着之前的状态取值，如此便可以取出正确的值，这种特性类似于弹匣，打枪的时候，子弹按顺序一个个的打出，不会出现混乱，扣动扳机就是执行一次迭代操作，打出子弹即是取出了一个值。

生成器也是迭代器的一种，迭代器只能记住自身的执行状态并等待下一次迭代，而生成器除了也会记住执行状态，还可以通过yield语句控制使多个生成器切换执行，例如手枪只能由一个枪口打出子弹，而加特林机枪可以通过旋转控制（yield切换）用多个枪口依次出子弹，威力也会更强，生成器是实现异步协程的重要基础。
