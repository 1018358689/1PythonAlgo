## 数学游戏，24点：任意输入4个1~10的数字，求所有四则运算结果为24的算式。

* 符号为O 数字为N 那么有 NONONON 例如 4*2+3/1 其中每个N有<=4个可能(不可重复) 每个O有四个可能(可重复) 
* 单数字4个位置位置排列组合有 4*3*2*1=24种 但是如四个初始数字有重复项会造成 排列重复 例如(6,6,1,3)和(6,6,1,3)是一样的 所以使用set()去冗余。最后会<=24种
* 数字间3个间隔插入符号排列 每个位置4种可能 共4*4*4=64种可能
* NONONON 排列组合完毕，需要考虑符号优先级插入括号，直接枚举出如下情况 其中132和312是一样的
* 计算式都举出来了，利用eval()计算 判断24
* 排列问题 A(n,r) = n*(n-1)*(n-2)*...*(n-r+1) 共r个相乘
* 组合问题 C(n,r) = A(n,r)/r!  =  (n*(n-1)*(n-2)*...*(n-r+1))/(r*(r-1)*...*1)  r个相乘 除以 r的阶乘

```python
from itertools import permutations
from itertools import product
def allin(numbers):
    data_list = set([i for i in permutations(numbers,r = 4)]) # 排列问题 A(set,r) 集合中取r个元素的排列 共A(4,4)=24种 set去重
    op_list = [i for i in product('+-*/', repeat = 3)] # 笛卡尔积  AXB = ([A1,B1],[A1,B2]...) 共C(4,1)**3=64种
    res = []
    for d in data_list:
        for o in op_list:
            c123 = '(({}{}{}){}{}){}{}'.format(d[0], o[0], d[1], o[1], d[2], o[2], d[3])
            c132 = '({}{}{}){}({}{}{})'.format(d[0], o[0], d[1], o[1], d[2], o[2], d[3])
            c213 = '({}{}({}{}{})){}{}'.format(d[0], o[0], d[1], o[1], d[2], o[2], d[3])
            c231 = '{}{}(({}{}{}){}{})'.format(d[0], o[0], d[1], o[1], d[2], o[2], d[3])
            c321 = '{}{}({}{}({}{}{}))'.format(d[0], o[0], d[1], o[1], d[2], o[2], d[3])
            cs = [c123, c132, c213, c231, c321]
            for c in cs:
                try:
                    if eval(c) == 24:
                        res.append(c)
                except:
                    pass
    return '没有24点！' if res == [] else res

allin([3, 8, 1, 9])
```
