在神经网络中，relu作为神经元的常用激活函数
请实现relu函数，其功能为输入一个数，输出relu值。

示例 1:
输入: 5
输出: 5

示例 2:
输入: 0
输出: 0

示例 3:
输入: -9
输出: 0

```python
def f(num):
    if num > 0:
        return num
    else:
        return 0


f(5), f(0), f(-5)
```

```python
def f(num):
    return num if num > 0 else 0


f(5), f(0), f(-5)
```

```python
def f(num):
    return num > 0 and num or 0


f(5), f(0), f(-5)
```

```python
def f(num):
    return max(0, num)


f(5), f(0), f(-5)
```
