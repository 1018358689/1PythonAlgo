11.17.四数相加II
给定四个包含整数的数组列表 A , B , C , D ,计算有多少个元组 (i, j, k, l) ，使得 A[i] + B[j] + C[k] + D[l] = 0。
为了使问题简单化，所有的 A, B, C, D 具有相同的长度 N，且 0 ≤ N ≤ 500 。所有整数的范围在 -2^28 到 2^28 - 1 之间，最终结果不会超过 2^31 - 1 。
例如:
输入: A = [ 1, 2] B = [-2,-1] C = [-1, 2] D = [ 0, 2] 输出: 2 解释: 两个元组如下:
1.(0, 0, 0, 1) -> A[0] + B[0] + C[0] + D[1] = 1 + (-2) + (-1) + 2 = 0
2.(1, 1, 0, 0) -> A[1] + B[1] + C[0] + D[0] = 2 + (-1) + (-1) + 0 = 0

```python
def f(A, B, C, D):
    cnt = 0
    for a in A:
        for b in B:
            for c in C:
                for d in D:
                    if not a + b + c + d: cnt += 1
    return cnt


f([1, 2], [-2, -1], [-1, 2], [0, 2])
```

```python
def f(A, B, C, D):
    dicABC = {}
    cnt = 0
    for a in A:
        for b in B:
            for c in C:
                s = a + b + c
                if s not in dicABC: dicABC[s] = 0
                dicABC[s] += 1
    for d in D:
        if -d in dicABC: cnt += dicABC[-d]
    return cnt


f([1, 2], [-2, -1], [-1, 2], [0, 2])
```

```python
def f(A, B, C, D):
    dicAB = {}
    cnt = 0
    for a in A:
        for b in B:
            s = a + b
            if s not in dicAB: dicAB[s] = 0
            dicAB[s] += 1
    for c in C:
        for d in D:
            if -c - d in dicAB: cnt += dicAB[-c - d]
    return cnt


f([1, 2], [-2, -1], [-1, 2], [0, 2])
```

```python
from collections import defaultdict
def f(A, B, C, D):
    dicAB = defaultdict(int)
    cnt = 0
    for a in A:
        for b in B:
            dicAB[a+b] += 1
    for c in C:
        for d in D:
            if -c - d in dicAB: cnt += dicAB[-c - d]
    return cnt


f([1, 2], [-2, -1], [-1, 2], [0, 2])
```

```python
def f(A, B, C, D):
    dicAB = {}
    cnt = 0
    for a in A:
        for b in B:
            s = a + b
            dicAB[s] = dicAB.get(s, 0) + 1
    for c in C:
        for d in D:
            cnt += dicAB.get(-c - d, 0)
    return cnt


f([1, 2], [-2, -1], [-1, 2], [0, 2])
```

```python
from collections import Counter


def f(A, B, C, D):
    countAB = Counter(a + b for a in A for b in B)  # 建一个a+b的哈希表
    cnt = 0  # 用来计数
    for c in C:  #遍历-c-d，为了找使得a+b+c+d=0
        for d in D:
            cnt += countAB.get(-c - d, 0)  # 看看-c-d是否在coutAB哈希表里
    return cnt


f([1, 2], [-2, -1], [-1, 2], [0, 2])
```

```python
from collections import Counter

def f(A, B, C, D):
    countAB = Counter(a + b for a in A for b in B)  # 建一个a+b的哈希表
    return sum(countAB.get(-c - d, 0) for c in C for d in D)  # 看看-c-d是否在coutAB哈希表里

f([1, 2], [-2, -1], [-1, 2], [0, 2])
```

```python
from collections import Counter
from itertools import product


def f(A, B, C, D):
    d1 = Counter(sum(n) for n in product(A, B))
    return sum(d1.get(-sum(n), 0) for n in product(C, D))

f([1, 2], [-2, -1], [-1, 2], [0, 2])
```
