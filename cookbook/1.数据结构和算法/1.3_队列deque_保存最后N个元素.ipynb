# 对最后几项记录做一个有限的历史记录统计。

## 保存有限的“历史记录”可算是collections.deque队列完美应用场景。

```python
# 对一系列文本做匹配，保留符合要求的文本。
from collections import deque

data = [
    '0aA', '1b', '2cA', '3d', '4eA', '5fA',
    '6gA', '7hA', '8iA', '9j', '10k'
]


def search(data, pattern, history=3):
    res = deque(maxlen=history)
    for d in data:
        if pattern in d:
            res.append(d)
            yield d,res
# 对于变量来说，赋值是复制值。但是对于列表来说，赋值其实只是赋值引用
# 对于序列来说，赋值是两个一起改变，不管是修改源列表还是目标列表

list(search(data, 'A', 4))
```

# deque(maxlen=N)创建固定长度队列。当有新纪录加入而队列已满时会自动移除最老的那条记录。

```python
from collections import deque
dq = deque(range(1, 11), maxlen=5)  # 最多容纳五个元素，1到10依次入队 1到5就满了 依次移除最老的记录 保留倒数五个
dq
```

```python
dq = deque(range(1,11)) 
dq.rotate(3) # 从尾部3个元素旋转到首部
dq
```

```python
dq = deque(range(1,11)) 
dq.rotate(-3) # 从头部3个元素旋转到尾部
dq
```
